–°—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∞—è –æ—Ü–µ–Ω–∫–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω–æ–≥–æ –º–æ–º–µ–Ω—Ç–∞: $\mu_{xy} = \overline{xy} - \overline{x} * \overline{y}$

–í—ã–≤–æ–¥: –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ –ø—Ä—è–º—ã–µ –ø–æ–∫–∞–∑—ã–≤–∞—é—Ç –æ—Å–Ω–æ–≤–Ω–æ–µ –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏–µ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω–æ–π —Å–≤—è–∑–∏ - –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–µ: –±–æ–ª–µ–µ –≤—ã—Å–æ–∫–∏–º –∑–Ω–∞—á–µ–Ω–∏—è–º –æ–¥–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—Ç –±–æ–ª–µ–µ –≤—ã—Å–æ–∫–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è –¥—Ä—É–≥–æ–π.
###–ü—É–Ω–∫—Ç 8
–í—ã—á–∏—Å–ª–∏—Ç—å –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω—ã–µ –æ—Ç–Ω–æ—à–µ–Ω–∏—è ùúÇùë¶ùë• –∏ ùúÇùë•ùë¶. –£–±–µ–¥–∏—Ç—å—Å—è, —á—Ç–æ
–Ω–µ—Ä–∞–≤–µ–Ω—Å—Ç–≤–∞ ùúÇùë¶ùë• > |ùëüùë•ùë¶| –∏ ùúÇùë•ùë¶ > |ùëüùë•ùë¶| –≤—ã–ø–æ–ª–Ω—è—é—Ç—Å—è. 

```python
#8
y_s = np.zeros(8)
x_s = np.zeros(8)
yy_s = 0
xx_s = 0
yx_f = 0
xy_f = 0
rel_y = 0
rel_x = 0
dispx=3659.7421992426875

for j in range(8):
    for i in range(8):
        y_s[j] += chast[i][j]*sr[i]
        x_s[j] += chast[j][i]*srx1[i]
for i in range(8):
    y_s[i] /= nx[i]
    x_s[i] /= ny[i]
    yy_s += nx[i]*y_s[i]
    xx_s += ny[i]*x_s[i]
yy_s /= n
xx_s /= n
for i in range(8):
    yx_f += nx[i]*((y_s[i] - yy_s)**2)
    xy_f += ny[i]*((x_s[i] - xx_s)**2)
yx_f /= n
xy_f /= n
rel_y = (yx_f/disp1)**0.5
rel_x = (xy_f/dispx)**0.5

print("–ö–æ—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏:", r)
print("–í—ã–±–æ—Ä–æ—á–Ω–æ–µ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω–æ–µ –æ—Ç–Ω–æ—à–µ–Ω–∏–µ Y –∏ X:", rel_y)
print('–í—ã–±–æ—Ä–æ—á–Ω–æ–µ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω–æ–µ –æ—Ç–Ω–æ—à–µ–Ω–∏–µ X –∏ Y:',rel_x)
if ((rel_y > r) and (rel_x > r)): print("–ù–µ—Ä–∞–≤–µ–Ω—Å—Ç–≤–∞ –≤—ã–ø–æ–ª–Ω—è—é—Ç—Å—è")
```

###–ü—É–Ω–∫—Ç 9
–î–ª—è –∑–∞–¥–∞–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω—É—é –∫—Ä–∏–≤—É—é –ø–∞—Ä–∞–±–æ–ª–∏—á–µ—Å–∫–æ–≥–æ –≤–∏–¥–∞ ùë¶ = ùõΩ0+ùõΩ1ùë•+ùõΩ2ùë•2. –û—Ç–æ–±—Ä–∞–∑–∏—Ç—å –ø–æ–ª—É—á–µ–Ω–Ω—É—é –∫—Ä–∏–≤—É—é –Ω–∞ –º–Ω–æ–∂–µ—Å—Ç–≤–µ –≤—ã–±–æ—Ä–∫–∏.

```python
cx,cx2,cx3,cx4,cy,cxy,cx2y=0,0,0,0,0,0,0
koor=[0]*int(597-320)
ltr=[0]*int(597-320)
for i in range(93):
    cx=cx+selectx[i]
    cx2=cx2+selectx[i]**2
    cx3=cx3+selectx[i]**3
    cx4=cx4+selectx[i]**4
    cy=cy+selecty[i]
    cxy=cxy+(selectx[i]*selecty[i])
    cx2y=cx2y+(selectx[i]**2*selecty[i])
l=np.array([[len(selectx),cx,cx2],[cx,cx2,cx3],[cx2,cx3,cx4]])
r=np.array([cy,cxy,cx2y])
vsp=np.linalg.inv(l).dot(r)
for i in range(len(selectx)):
    plt.plot(selectx[i],selecty[i],'g.')
for i in range(int(597-320)):
    koor[i]=i+320
    ltr[i]=vsp[0]+vsp[1]*(i+320)+vsp[2]*(i+320)**2
plt.plot(koor,ltr)
plt.grid()
plt.show()
print(vsp)
```

###–ü—É–Ω–∫—Ç 11
–í—ã—á–∏—Å–ª–∏—Ç—å —Å–ª–µ–¥—É—é—â–∏–µ –ø–æ–∫–∞–∑–∞—Ç–µ–ª–∏ –∫–∞—á–µ—Å—Ç–≤–∞ —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ
–∏–∑ —á–µ—Ç—ã—Ä—ë—Ö —É—Ä–∞–≤–Ω–µ–Ω–∏–π.
