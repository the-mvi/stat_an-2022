```python
def create_interval_series(sample):
    k = int(round(1 + np.log2(len(sample)))) # —Ñ–æ—Ä–º—É–ª–∞ –°—Ç—ë—Ä–¥–∂–µ—Å—Å–∞
    R = int(sample.max() - sample.min()) # —Ä–∞–∑–º–∞—Ö –≤—ã–±–æ—Ä–∫–∏
    h = float(R / k) # –¥–ª–∏–Ω–∞ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞
    
    i = []
    intervals = []
    centers = []
    n = []
    p = []
    x_0 = int(sample.min())
    
    for j in range(k):
        i.append(j + 1)
        
        interval_borders = [x_0, x_0 + h]
        center = (x_0 + x_0 + h) / 2
        interval_mask = (sample.values >= x_0) & (sample.values <= x_0 + h)
        interval = sample.loc[interval_mask]
        n_i = len(interval)
        p_i = n_i / len(sample)
        
        intervals.append(interval_borders)
        centers.append(center)
        n.append(n_i)
        p.append(p_i)
        
        x_0 = x_0 + h
        
    Data = {'$i$': i, '$(x_{i-1}, x_{i}]$': intervals, '$\widetilde{x_{i}}$': centers, '$n_{i}$': n, '$p_{i}^*$': p}
    data_frame = pd.DataFrame(data=Data)
    
    return data_frame, h
```

```python
interval_series, step = create_interval_series(sample_sorted)
interval_series.style.hide_index()
```

```python
def task5(int_centers, sample, int_series):
    intervals = int_series['$(x_{i-1}, x_{i}]$']
    accum_freqs = []
    accum_rel_freqs = []
    
    for i in range(len(int_centers)):
        mask = (sample.values <= intervals.values[i][1])
        values = sample.loc[mask]
        accum_freqs.append(len(values))
        accum_rel_freqs.append(len(values) / len(sample))
        
    table_cols = {'$n_{i}^{\Sigma}$': accum_freqs, '$p_{i}^{\Sigma}$': accum_rel_freqs}
    table = pd.DataFrame(data=table_cols) # –Ω–∞–∫–æ–ø–ª–µ–Ω–Ω—ã–µ –∞–±—Å–æ–ª—é—Ç–Ω—ã–µ –∏ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω—ã–µ —á–∞—Å—Ç–æ—Ç—ã
    
    res = pd.concat([int_series, table], axis=1)
    
    freqs = int_series['$n_{i}$']
    rel_freqs = int_series['$p_{i}^*$']
    sum_row = {'$i$': ['$\Sigma$'], '$(x_{i-1}, x_{i}]$': ['-'], '$\widetilde{x_{i}}$': ['-'], '$n_{i}$': [freqs.sum()], '$p_{i}^*$': [rel_freqs.sum()], '$n_{i}^{\Sigma}$': ['-'], '$p_{i}^{\Sigma}$': ['-']}
    sum_tab = pd.DataFrame(data=sum_row)
    
    res = pd.concat([res, sum_tab], axis=0)
    
    return res
```

<h1>–ó–∞–¥–∞–Ω–∏–µ 6</h1>

```python
def cond_variant(var, C, h):
    return (var - C) / h
```

```python
def task6(tab_from_task5, Step):
    i = []
    x = tab_from_task5['$\widetilde{x_{i}}$'].values[0:len(tab_from_task5['$\widetilde{x_{i}}$'].values)-1]
    n = tab_from_task5['$n_{i}$'].values[0:len(tab_from_task5['$n_{i}$'].values)-1]
    u = []
    nu = []
    nu2 = []
    nu3 = []
    nu4 = []
    n_u_plus_1_4 = []
    
    C = x.max()
    
    for j in range(len(x)):
        i.append(j + 1)
        u_i = cond_variant(x[j], C, Step)
        u.append(u_i)
        
        nu.append(n[j] * u_i)
        nu2.append(n[j] * u_i**2)
        nu3.append(n[j] * u_i**3)
        nu4.append(n[j] * u_i**4)
        n_u_plus_1_4.append(n[j] * (u_i + 1)**4)
    
    table = {'$i$': i, '$x_{i}$': x, '$n_{i}$': n, '$u_{i}$': u, '$n_{i} u_{i}$': nu, '$n_{i} u_{i}^2$': nu2, '$n_{i} u_{i}^3$': nu3, '$n_{i} u_{i}^4$': nu4, '$n_{i} (u_{i} + 1)^4$': n_u_plus_1_4}
    table_df = pd.DataFrame(data=table)
    
    sum_row = {'$i$': ['$\Sigma$'], '$x_{i}$': ['-'], '$n_{i}$': [n.sum()], '$u_{i}$': ['-'], '$n_{i} u_{i}$': [sum(nu)], '$n_{i} u_{i}^2$': [sum(nu2)], '$n_{i} u_{i}^3$': [sum(nu3)], '$n_{i} u_{i}^4$': [sum(nu4)], '$n_{i} (u_{i} + 1)^4$': [sum(n_u_plus_1_4)]}
    sum_tab = pd.DataFrame(data=sum_row)
    
    res = pd.concat([table_df, sum_tab], axis=0)
    
    if (sum(n_u_plus_1_4) == sum(nu4) + 4 * sum(nu3) + 6 * sum(nu2) + 4 * sum(nu) + n.sum()): print('–ü—Ä–æ–≤–µ—Ä–∫–∞ –ø—Ä–æ–π–¥–µ–Ω–∞')
    else: print('–ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–µ –ø—Ä–æ–π–¥–µ–Ω–∞')
    
    return res, C
```

```python
def task7(tab_from_t6, Step):
    n = tab_from_t6['$n_{i}$'].values[len(tab_from_t6['$n_{i}$'].values) - 1]
    nu_sum = tab_from_t6['$n_{i} u_{i}$'].values[len(tab_from_t6['$n_{i} u_{i}$'].values) - 1]
    nu2_sum = tab_from_t6['$n_{i} u_{i}^2$'].values[len(tab_from_t6['$n_{i} u_{i}^2$'].values) - 1]
    nu3_sum = tab_from_t6['$n_{i} u_{i}^3$'].values[len(tab_from_t6['$n_{i} u_{i}^3$'].values) - 1]
    nu4_sum = tab_from_t6['$n_{i} u_{i}^4$'].values[len(tab_from_t6['$n_{i} u_{i}^4$'].values) - 1]
    
    cond_emp_moments = [nu_sum / n, nu2_sum / n, nu3_sum / n, nu4_sum / n]
    
    mu1 = 0
    mu2 = (cond_emp_moments[1] - cond_emp_moments[0]**2) * Step**2
    mu3 = (cond_emp_moments[2] - 3 * cond_emp_moments[1] * cond_emp_moments[0] + 2 * cond_emp_moments[0]**3) * Step**3
    mu4 = (cond_emp_moments[3] - 4 * cond_emp_moments[2] * cond_emp_moments[0] + 6 * cond_emp_moments[1] * cond_emp_moments[0]**2 - 3 * cond_emp_moments[0]**4) * Step**4
    
    cent_emp_moments = [mu1, mu2, mu3, mu4]
    
    table = {'$i$': [1, 2, 3, 4], '$\nu_{i}^*$': cond_emp_moments, '$\mu_{i}^*$': cent_emp_moments}
    res = pd.DataFrame(data=table)
    
    return res
```

```python
def get_modal_interval(interval_ser):
    freqs = interval_ser['$n_{i}$'].values
    max_val = freqs.max()
    
    intervals = interval_ser['$(x_{i-1}, x_{i}]$'].values
    
    for i in range(len(freqs)):
        if freqs[i] == max_val:
            if i != 0 and i != len(freqs) - 1:
                return intervals[i], freqs[i], freqs[i - 1], freqs[i + 1]
            elif i == 0:
                return intervals[i], freqs[i], 0, freqs[i + 1]
            elif i == len(freqs) - 1:
                return intervals[i], freqs[i], freqs[i - 1], 0
```

```python
from scipy.stats import norm
```

```python
def check_hypothesis(interv_ser, a, sigma):
    j = []
    n_i = interv_ser['$n_{i}$'].values
    intervals = interv_ser['$(x_{i-1}, x_{i}]$'].values
    centers = interv_ser['$\widetilde{x_{i}}$'].values
    
    p_i = []
    N_i = [] # –í—ã—Ä–∞–≤–Ω–∏–≤–∞—é—â–∏–µ —á–∞—Å—Ç–æ—Ç—ã
    ni_minus_Ni_square = []
    ni_minus_Ni_Ni = []
    ni2 = []
    ni2_Ni = []
    
    Phi = lambda x: erf(x / 2**0.5) / 2
    
    for i in range(len(intervals)):
        j.append(i + 1)
        first = intervals[i][0]
        last = intervals[i][1]
        p_i_val = 0
        if i != 0 and i != len(intervals):
            p_i_val = round(norm.cdf((last - a) / sigma), 4) - round(norm.cdf((first - a) / sigma), 4)
        elif i == 0:
            p_i_val = round(norm.cdf((last - a) / sigma), 4)
        elif i == len(intervals):
            p_i_val = 1 - round(norm.cdf((first - a) / sigma), 4)
        p_i.append(p_i_val)
        N_i_val = n_i.sum() * p_i_val
        N_i.append(N_i_val)
        
        ni_minus_Ni_square.append((n_i[i] - N_i_val)**2)
        ni_minus_Ni_Ni.append((n_i[i] - N_i_val)**2 / N_i_val)
        ni2.append(n_i[i]**2)
        ni2_Ni.append(n_i[i]**2 / N_i_val)
    
    if sum(ni_minus_Ni_Ni) != (sum(ni2_Ni) - n_i.sum()):
        print("–ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–µ –ø—Ä–æ–π–¥–µ–Ω–∞")
    
    table_data = {'$i$': j, '$(x_{i-1}, x_{i}]$': intervals, '$n_{i}$': n_i, '$p_{i}$': p_i, '$n_{i}^{\'}$': N_i, '$(n_{i} - n_{i}^{\'})^2$': ni_minus_Ni_square, '$\\frac{(n_{i} - n_{i}^{\'})^2}{n_{i}^{\'}} $': ni_minus_Ni_Ni, '$n_{i}^2$': ni2, '$\\frac{n_{i}^2}{n_{i}^{\'}}$': ni2_Ni}
    table_df = pd.DataFrame(data=table_data)
    
    sum_row = {'$i$': ['$\Sigma$'], '$(x_{i-1}, x_{i}]$': ['-'], '$n_{i}$': [n_i.sum()], '$p_{i}$': [sum(p_i)], '$n_{i}^{\'}$': [sum(N_i)], '$(n_{i} - n_{i}^{\'})^2$': ['-'], '$\\frac{(n_{i} - n_{i}^{\'})^2}{n_{i}^{\'}} $': [sum(ni_minus_Ni_Ni)], '$n_{i}^2$': ['-'], '$\\frac{n_{i}^2}{n_{i}^{\'}}$': [sum(ni2_Ni)]}
    sum_tab = pd.DataFrame(data=sum_row)
    
    res_df = pd.concat([table_df, sum_tab], axis=0)
    display(res_df.style.hide_index())
    
    # alpha = 0.05
    # df = k - r - 1 = 8 - 2 - 1 = 5
    # –∫—Ä–∏—Ç–∏—á–µ—Å–∫–∞—è –≤–µ–ª–∏—á–∏–Ω–∞ —Ä–∞–≤–Ω–∞ 11.1
    
    crit_val = 11.1
    
    statistics = sum(ni_minus_Ni_Ni)
    
    if statistics < crit_val: print("–≥–∏–ø–æ—Ç–µ–∑—É –ø—Ä–∏–Ω–∏–º–∞–µ–º")
    else: print("–ì–∏–ø–æ—Ç–µ–∑—É –æ—Ç—Ç–æ—Ä–≥–∞–µ–º")
```
