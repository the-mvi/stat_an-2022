## Основные теоретические положения
<p>Генеральная совокупность - совокупность объектов, из которых производится выборка. Объём = N</p>
<p>Выборка - совокупность случайно отобранных объектов. Объём = n</p>
<p>Объём выборки - количество элементов в выборке.</p>
<p>Виды выборок: простая, механическая, типическая, серийная.</p>
<p>Ранжированный ряд - расположение единиц совокупности в порядке возрастания или убывания выбранного признака</p>
<p>Вариационный ряд - ранжированный в порядке возрастания или убывания ряд вариант с соответствующими им весами</p>
<p>Интервальный ряд - соответствие между интервалом изменения признака и числом объектов, попадающих в данный интервал</p>
<p>Полигон частот - ломаная линия, соединяющая точки дискретного ряда $(x_{1}; n_{1}), ... , (x_{k}; n_{k})$ </p>
<p>Полигон относительных частот - ломаная линия, соединяющая точки дискретного ряда $(x_{1}; p_{1}^*), ... , (x_{k}; p_{k}^*)$</p>
<p>Гистограмма частот - ступенчатая фигура, состоящая из прямоугольников, основаниями которых являются частичные интервалы, а высоты равны $f_{i} = \frac{n_{i}}{h}$, где $h$ - длина интервала</p>
<p>Гистограмма относительных частот - ступенчатая фигура, состоящая из прямоугольников, основаниями которых являются частичные интервалы, а высоты равны $f_{i}^* = \frac{p_{i}^*}{h}$, где $h$ - длина интервала</p>
<p>Эмпирическая функция - функция $F^{*}(x)$, задающая для каждого значения $x$ относительную частоту события ${X < x}$</p>
<p>Накопленная абсолютная частота $n_{i}^{\Sigma}$ - число наблюдений, при которых наблюдалось значение признака, меньше чем $x_{i}$</p>
<p>Накопленная относительная частота $p_{i}^{\Sigma}$ - отношение накопленной частоты к общему числу наблюдений</p>
<p>Условные варианты: $u_{i} = \frac {x_{i} - C} {h}$, где $C$ - ложный нуль, $h$ - длина интервала</p>
<p>Условный эмпирический момент порядка $r$: $\nu_{r}^* = \frac{1}{n} \sum_{i=1}^k {n_{i}} (\frac {x_{i} - C} {h})^r$</p>
<p>Центральные моменты через условные эмпирические моменты:</p>
<p>$\mu_{1}^* = 0$</p>
<p>$\mu_{2}^* = (\nu_{2}^* - (\nu_{1}^*)^2)h^2$</p>
<p>$\mu_{3}^* = (\nu_{3}^* - 3\nu_{2}^* \nu_{1}^* + 2(\nu_{1}^*)^3)h^3$</p>
<p>$\mu_{4}^* = (\nu_{4}^* - 4\nu_{3}^* \nu_{1}^* + 6\nu_{2}^* (\nu_{1}^*)^2 - 3(\nu_{1}^*)^4)h^4$</p>
<p>Выборочное среднее: $\overline{x_{в}} = \frac {1}{n} \sum_{i = 1}^{k} x_{i}n_{i} = \sum_{i=1}^{k} x_{i}p_{i}^*$ </p>
<p>Выборочная дисперсия: $\sigma_{в}^2 = \frac {1}{n} \sum_{i=1}^{k} n_{i}(x_{i} - \overline{x_{в}})^2 = \sum_{i=1}^{k} p_{i}^*(x_{i} - \overline{x_{в}})^2$</p>
<p>Выборочное СКО: $\sigma_{в} = \sqrt{\sigma_{в}^2}$</p>
<p>Выборочное среднее с помощью условных вариант: $\overline{x_{в}} = h \nu_{1}^* + C$</p>
<p>Выборочная дисперсия с помощью условных вариант: $\sigma_{в}^2 = \mu_{2}^*$</p>
<p>Исправленная дисперсия: $\sigma_{в}^2 \frac{n}{n - 1}$</p>
<p>Исправленное СКО - корень из исправленной дисперсии.</p>
<p>Коэффициент асимметрии: $a_{s}^* = \frac {\mu_{3}^*}{\sigma_{в}^3}$</p>
<p>Коэффициент эксцесса: $\varepsilon_{k}^* = \frac {\mu_{4}^*}{\sigma_{в}^4} - 3$</p>
<p>Мода - значение во множестве наблюдений, которое встречается наиболее часто</p>
<p>$Mo^* = x_{Mo}^(0) + h \frac {n_{Mo} - n_{Mo - 1}} {(n_{Mo} - n_{Mo - 1}) + (n_{Mo} - n_{Mo + 1})}$</p>
<p>Медиана - число, которое находится в середине этого набора</p>
<p>$Me^* = x_{Me}^(0) + \frac{h}{p_{Me}^*} (0.5 - p_{Me - 1}^{\Sigma})$</p>
<p>Коэффициент вариации: $V^* = \frac {\sigma_{в}}{|\overline{x_{в}}|} \cdot 100%$</p>
<p>Доверительный интервал для математического ожидания при неизвестном среднеквадратичном отклонении при заданном объёме выборки:</p>
<p>$(\overline{x_{в}} - \frac {t_{\gamma} \sigma}{\sqrt{n}}; \overline{x_{в}} + \frac {t_{\gamma} \sigma}{\sqrt{n}})$</p>
<p>Точность: $\frac {t_{\gamma} \sigma}{\sqrt{n}}$</p>
<p>Доверительный интервал для среднеквадратичного отклонения:</p>
<p>$s(1 - q) < \sigma_{г} < s(1 + q), q < 1$</p>
<p>$0 < \sigma_{г} < s(1 + q), q > 1$</p>
<p>Теоритические частоты: $P(x_{i - 1} < x < x_{i}) = \Phi(\frac{x_{i} - a}{\sigma}) - \Phi(\frac{x_{i-1} - a}{\sigma})$</p>
<p>Статистика: $\chi_{набл}^2 = \sum_{i=1}^{k} \frac{(n_{i} - n_{i}^{'})^2}{n_{i}^{'}} = \sum_{i=1}^{k} \frac{n_{i}^2}{n_{i}^{'}} - n$</p>
<p>Критическая величина: $\chi_{крит}^2 = \chi^2(\alpha, df), df = k - r - 1$, где $k$ - число интервалов, $r$ - число параметров дифференциальной функции распределения</p>
<p>Условие принятия гипотезы: $\chi_{набл}^2 < \chi_{крит}^2$</p>

<p>Интервальный ряд</p>

```python
def create_interval_series(sample):
    k = int(round(1 + np.log2(len(sample)))) # формула Стёрджесса
    R = int(sample.max() - sample.min()) # размах выборки
    h = float(R / k) # длина интервала
    
    i = []
    intervals = []
    centers = []
    n = []
    p = []
    x_0 = int(sample.min())
    
    for j in range(k):
        i.append(j + 1)
        
        interval_borders = [x_0, x_0 + h]
        center = (x_0 + x_0 + h) / 2
        interval_mask = (sample.values >= x_0) & (sample.values <= x_0 + h)
        interval = sample.loc[interval_mask]
        n_i = len(interval)
        p_i = n_i / len(sample)
        
        intervals.append(interval_borders)
        centers.append(center)
        n.append(n_i)
        p.append(p_i)
        
        x_0 = x_0 + h
        
    Data = {'$i$': i, '$(x_{i-1}, x_{i}]$': intervals, '$\widetilde{x_{i}}$': centers, '$n_{i}$': n, '$p_{i}^*$': p}
    data_frame = pd.DataFrame(data=Data)
    
    return data_frame, h
```

```python
interval_series, step = create_interval_series(sample_sorted)
interval_series.style.hide_index()
```

```python
def task5(int_centers, sample, int_series):
    intervals = int_series['$(x_{i-1}, x_{i}]$']
    accum_freqs = []
    accum_rel_freqs = []
    
    for i in range(len(int_centers)):
        mask = (sample.values <= intervals.values[i][1])
        values = sample.loc[mask]
        accum_freqs.append(len(values))
        accum_rel_freqs.append(len(values) / len(sample))
        
    table_cols = {'$n_{i}^{\Sigma}$': accum_freqs, '$p_{i}^{\Sigma}$': accum_rel_freqs}
    table = pd.DataFrame(data=table_cols) # накопленные абсолютные и относительные частоты
    
    res = pd.concat([int_series, table], axis=1)
    
    freqs = int_series['$n_{i}$']
    rel_freqs = int_series['$p_{i}^*$']
    sum_row = {'$i$': ['$\Sigma$'], '$(x_{i-1}, x_{i}]$': ['-'], '$\widetilde{x_{i}}$': ['-'], '$n_{i}$': [freqs.sum()], '$p_{i}^*$': [rel_freqs.sum()], '$n_{i}^{\Sigma}$': ['-'], '$p_{i}^{\Sigma}$': ['-']}
    sum_tab = pd.DataFrame(data=sum_row)
    
    res = pd.concat([res, sum_tab], axis=0)
    
    return res
```

```python
def cond_variant(var, C, h):
    return (var - C) / h
```

```python
def task6(tab_from_task5, Step):
    i = []
    x = tab_from_task5['$\widetilde{x_{i}}$'].values[0:len(tab_from_task5['$\widetilde{x_{i}}$'].values)-1]
    n = tab_from_task5['$n_{i}$'].values[0:len(tab_from_task5['$n_{i}$'].values)-1]
    u = []
    nu = []
    nu2 = []
    nu3 = []
    nu4 = []
    n_u_plus_1_4 = []
    
    C = x.max()
    
    for j in range(len(x)):
        i.append(j + 1)
        u_i = cond_variant(x[j], C, Step)
        u.append(u_i)
        
        nu.append(n[j] * u_i)
        nu2.append(n[j] * u_i**2)
        nu3.append(n[j] * u_i**3)
        nu4.append(n[j] * u_i**4)
        n_u_plus_1_4.append(n[j] * (u_i + 1)**4)
    
    table = {'$i$': i, '$x_{i}$': x, '$n_{i}$': n, '$u_{i}$': u, '$n_{i} u_{i}$': nu, '$n_{i} u_{i}^2$': nu2, '$n_{i} u_{i}^3$': nu3, '$n_{i} u_{i}^4$': nu4, '$n_{i} (u_{i} + 1)^4$': n_u_plus_1_4}
    table_df = pd.DataFrame(data=table)
    
    sum_row = {'$i$': ['$\Sigma$'], '$x_{i}$': ['-'], '$n_{i}$': [n.sum()], '$u_{i}$': ['-'], '$n_{i} u_{i}$': [sum(nu)], '$n_{i} u_{i}^2$': [sum(nu2)], '$n_{i} u_{i}^3$': [sum(nu3)], '$n_{i} u_{i}^4$': [sum(nu4)], '$n_{i} (u_{i} + 1)^4$': [sum(n_u_plus_1_4)]}
    sum_tab = pd.DataFrame(data=sum_row)
    
    res = pd.concat([table_df, sum_tab], axis=0)
    
    if (sum(n_u_plus_1_4) == sum(nu4) + 4 * sum(nu3) + 6 * sum(nu2) + 4 * sum(nu) + n.sum()): print('Проверка пройдена')
    else: print('Проверка не пройдена')
    
    return res, C
```

```python
def get_modal_interval(interval_ser):
    freqs = interval_ser['$n_{i}$'].values
    max_val = freqs.max()
    
    intervals = interval_ser['$(x_{i-1}, x_{i}]$'].values
    
    for i in range(len(freqs)):
        if freqs[i] == max_val:
            if i != 0 and i != len(freqs) - 1:
                return intervals[i], freqs[i], freqs[i - 1], freqs[i + 1]
            elif i == 0:
                return intervals[i], freqs[i], 0, freqs[i + 1]
            elif i == len(freqs) - 1:
                return intervals[i], freqs[i], freqs[i - 1], 0
```

```python
def check_hypothesis(interv_ser, a, sigma):
    j = []
    n_i = interv_ser['$n_{i}$'].values
    intervals = interv_ser['$(x_{i-1}, x_{i}]$'].values
    centers = interv_ser['$\widetilde{x_{i}}$'].values
    
    p_i = []
    N_i = [] # Выравнивающие частоты
    ni_minus_Ni_square = []
    ni_minus_Ni_Ni = []
    ni2 = []
    ni2_Ni = []
    
    Phi = lambda x: erf(x / 2**0.5) / 2
    
    for i in range(len(intervals)):
        j.append(i + 1)
        first = intervals[i][0]
        last = intervals[i][1]
        p_i_val = 0
        if i != 0 and i != len(intervals):
            p_i_val = round(norm.cdf((last - a) / sigma), 4) - round(norm.cdf((first - a) / sigma), 4)
        elif i == 0:
            p_i_val = round(norm.cdf((last - a) / sigma), 4)
        elif i == len(intervals):
            p_i_val = 1 - round(norm.cdf((first - a) / sigma), 4)
        p_i.append(p_i_val)
        N_i_val = n_i.sum() * p_i_val
        N_i.append(N_i_val)
        
        ni_minus_Ni_square.append((n_i[i] - N_i_val)**2)
        ni_minus_Ni_Ni.append((n_i[i] - N_i_val)**2 / N_i_val)
        ni2.append(n_i[i]**2)
        ni2_Ni.append(n_i[i]**2 / N_i_val)
    
    if sum(ni_minus_Ni_Ni) != (sum(ni2_Ni) - n_i.sum()):
        print("Проверка не пройдена")
    
    table_data = {'$i$': j, '$(x_{i-1}, x_{i}]$': intervals, '$n_{i}$': n_i, '$p_{i}$': p_i, '$n_{i}^{\'}$': N_i, '$(n_{i} - n_{i}^{\'})^2$': ni_minus_Ni_square, '$\\frac{(n_{i} - n_{i}^{\'})^2}{n_{i}^{\'}} $': ni_minus_Ni_Ni, '$n_{i}^2$': ni2, '$\\frac{n_{i}^2}{n_{i}^{\'}}$': ni2_Ni}
    table_df = pd.DataFrame(data=table_data)
    
    sum_row = {'$i$': ['$\Sigma$'], '$(x_{i-1}, x_{i}]$': ['-'], '$n_{i}$': [n_i.sum()], '$p_{i}$': [sum(p_i)], '$n_{i}^{\'}$': [sum(N_i)], '$(n_{i} - n_{i}^{\'})^2$': ['-'], '$\\frac{(n_{i} - n_{i}^{\'})^2}{n_{i}^{\'}} $': [sum(ni_minus_Ni_Ni)], '$n_{i}^2$': ['-'], '$\\frac{n_{i}^2}{n_{i}^{\'}}$': [sum(ni2_Ni)]}
    sum_tab = pd.DataFrame(data=sum_row)
    
    res_df = pd.concat([table_df, sum_tab], axis=0)
    display(res_df.style.hide_index())
    
    # alpha = 0.05
    # df = k - r - 1 = 8 - 2 - 1 = 5
    # критическая величина равна 11.1
    
    crit_val = 11.1
    
    statistics = sum(ni_minus_Ni_Ni)
    
    if statistics < crit_val: print("гипотезу принимаем")
    else: print("Гипотезу отторгаем")
```
