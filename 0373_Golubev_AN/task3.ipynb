## Основные теоретические положения
<p>Нормализация элементов выборки производится следующим образом: $x^{'} = \frac{x - \overline{x_{в}}}{\sigma}$</p>
<p>$x^{'}$ - нормализованный элемент</p>
<p>$x$ - исходный элемент</p>
<p>$\overline{x_{в}}$ - выборочное среднее</p>
<p>$\sigma$ - выборочное СКО</p>
<br>

<p>Кластерный анализ - задача разбиения заданной выборки объектов на подмножества так, чтобы каждый кластер состоял из схожих объектов.</p>
<p><b>Этапы кластеризации:</b></p>
<p>1) Отбор выборки</p>
<p>1.2) Нормализация выборки (необязательный этап)</p>
<p>2) Определение множества переменных, по которым будут оцениваться объекты выборки</p>
<p>3) Вычисление значений меры сходства между объектами</p>
<p>4) Применение метода для создания групп</p>
<p>5) Предоставление результатов</p>
<p>6) Корректировка результатов</p>
<br>

<p>Метрика - вещественная функция $d(x, y)$, обладающая следующими свойствами:</p>
<p>1) $d(x, y) \geq 0$</p>
<p>2) $d(x, y) = d(y, x)$</p>
<p>3) $d(x, y) \leq d(x, z) + d(z, y)$</p>
<br>

<p>Примеры метрик:</p>
<p>Евклидово расстояние: $d(x, y) = \sqrt{\displaystyle\sum_{i = 1}^{n} {(x_i - y_i)^2}}$</p>
<p>Манхеттенское расстояние: $d(x, y) = \displaystyle\sum_{i = 1}^{n} {|x_i - y_i|}$</p>
<br>

<p>Метод K-means (используется Евклидово расстояние):</p>
<p>1) Выбрать число кластеров $k$</p>
<p>2) Случайным образом выбрать $k$ центроидов</p>
<p>3) Отнести каждый объект к кластеру с ближайшим центроидом</p>
<p>4) Пересчитать центроиды (вычислить средние значения кластеров)</p>
<p>5) Если новые и старые центроиды не совпадают, перейти к пункту 3</p>
<br>

<p>Метод K-medians - вариация метода K-means. Алгоритм аналогичен методу K-means, но:</p>
<p>1) В качестве метрики используется Манхеттенское расстояние</p>
<p>2) Вместо вычисления средних вычисляются медины</p>
<br>

<p>Верхняя оценка количества кластеров: $\widetilde{k} = \lfloor{\sqrt{n \over 2}}\rfloor$</p>
<br>

<p><b>Метод силуэтов для оценки оптимального количества кластеров:</b></p>
<p>Для каждого значения $k$:</p>
<p>1) Вычисляем кластеры, используя исследуемый метод</p>
<p>2) Для каждой точки в выборке $x_i \in C_i$ вычисляем коэффициент силуэта $s_i = \frac {b_i - a_i}{max(a_i; b_i)}$, где</p>
<p>$a_i = \frac{1}{|C_i| - 1} \displaystyle\sum_{x_j \in C_i, j \neq i} {d(x_j; x_i)}$ - внутрикластерное расстояние</p>
<p>$b_i = min \frac{1}{|C_j|} \displaystyle\sum_{x_j \in C_j} {d(x_j; x_i)}, j \neq i$ - среднее расстояние до ближайшего кластера</p>
<p>3)Вычисляем среднее значение коэффициента силуэтов для $k$</p>
<p>В качестве оптимального количества кластеров принимается $k$, для которого среднее значение коэффициента силуэтов максимально</p>
<br>

<p><b>Метод FOREL для поиска кластеров:</b></p>
<p>1) Задаём множество точек $U$ и радиус $R$</p>
<p>2) Случайным образом выбираем $x_c \in U$</p>
<p>3) Задаём множество $K$ := {$x \in U | d(x, x_c) \leq R$}</p>
<p>4) Вычисляем $\widetilde{x_c} := \frac{1}{|K|} \displaystyle\sum_{x \in K} {x}$</p>
<p>5) Если $x_c \neq \widetilde{x_c}$, то $x_c := \widetilde{x_c}$ и возвращаемся к пункту 3)</p>
<p>6) $U := U \diagdown K$</p>
<p>7) Добавляем $K$ к множеству кластеров, а $x_c$ к множеству центроидов</p>
<p>8) Если $|U| > 0$, то возвращаемся к пункту 2), иначе конец</p>
<br>

<p><b>Подбор радиуса для метода FOREL:</b></p>
<p>$D$ = {$d_{ij}$} - матрица расстояний</p>
<p>$d_{ij} = d(x_i; x_j)$</p>
<p>$R_{min}$ = min{$d_{ij} > 0$}</p>
<p>$R_{max}$ = max{$d_{ij}$}</p>
<p>$R_{min} \leq R \leq R_{max}$</p>
<br>

<p><b>Метод K-means++:</b></p>
<p>1) Случайным образом выбираем первый центроид среди точек выборки</p>
<p>2) Для каждой точки посчитать расстояние до ближайшего относительно неё центроида из уже имеющихся</p>
<p>3) Добавить новый центроид (точку, расстояние которой до ближайшего центроида максимально)</p>
<p>Повторять пункты 2) и 3), пока не будет получено $k$ центроидов</p>
<p>Далее для полученных кластеров используем стандартный K-means</p>

```python
def create_2D_normalized_sample(sampX, sampY):
    res = []
    for i in range(len(sampX)):
        res.append([sampX[i], sampY[i]])
    
    return res
```

```python
def Euclidian_dist(point1, point2):
    vec = [point1[0] - point2[0], point1[1] - point2[1]]
    return np.sqrt(vec[0]**2 + vec[1]**2)
```

```python
def K_means(k, sample, centroids):
    cur_centroids = copy.deepcopy(centroids)
    
    while True:
        prev_centroids = copy.deepcopy(cur_centroids)
        clusters = [ [] for _ in range(k)]
        
        for el in sample:
            dists = []
            for c in prev_centroids: dists.append(Euclidian_dist(el, c))
            index_min = min(range(len(dists)), key=dists.__getitem__)
            clusters[index_min].append(el)
        
        #for cl in clusters: print(cl)
        
        for i in range(k):
            cluster_i = clusters[i]
            x_vals = [el[0] for el in cluster_i]
            y_vals = [el[1] for el in cluster_i]
            
            cur_centroids[i] = [sum(x_vals) / len(x_vals), sum(y_vals) / len(y_vals)]
        
        if cur_centroids == prev_centroids: return clusters, cur_centroids
```

```python
def show_clusters(clusters, centroids, colors):
    fig, ax = plt.subplots()
    k = len(clusters)
    
    ax.grid(True)
    ax.set_title("k = " + str(k))
    ax.set_xlabel("X (Cocoa Percent)")
    ax.set_ylabel("Y (Rating)")
    
    for i in range(k):
        cluster_i = clusters[i]
        x_vals = [el[0] for el in cluster_i]
        y_vals = [el[1] for el in cluster_i]
        
        ax.plot(x_vals, y_vals, colors[i] + "o")
        ax.plot(centroids[i][0], centroids[i][1], colors[i] + "x", markersize=20)
```

<p>В методе K-means начальные центроиды можно выбрать случайным образом, но для простоты были взяты фиксированные точки.</p>

```python
def Manhattan_dist(point1, point2):
    return abs(point1[0] - point2[0]) + abs(point1[1] - point2[1])
```

```python
def median(samp):
    samp_sort = sorted(samp)
    if len(samp_sort) % 2 == 1:
        return samp_sort[len(samp_sort) // 2]
    else:
        return 0.5 * (samp_sort[len(samp_sort) // 2 - 1] + samp_sort[len(samp_sort) // 2])
```

```python
def K_medians(k, sample, centroids):
    cur_centroids = copy.deepcopy(centroids)
    
    while True:
        prev_centroids = copy.deepcopy(cur_centroids)
        clusters = [ [] for _ in range(k)]
        
        for el in sample:
            dists = []
            for c in prev_centroids: dists.append(Manhattan_dist(el, c))
            index_min = min(range(len(dists)), key=dists.__getitem__)
            clusters[index_min].append(el)
        
        #for cl in clusters: print(cl)
        
        for i in range(k):
            cluster_i = clusters[i]
            x_vals = [el[0] for el in cluster_i]
            y_vals = [el[1] for el in cluster_i]
            
            cur_centroids[i] = [np.median(np.array(x_vals)), np.median(np.array(y_vals))]
        
        if cur_centroids == prev_centroids: return clusters, cur_centroids
```

<p>В методе K-medians начальные центроиды можно выбрать случайным образом, но для простоты были взяты фиксированные точки.</p>

```python
for i in range(threshold):
    k_med_clusters, k_med_centroids = K_medians(i + 1, samp2D_normalized, init_centroids[i])
    show_clusters(k_med_clusters, k_med_centroids, colors)
```

```python
def silhouette_method(threshold, clustering_method, metric, sample, init_centroids):
    k_vals = []
    s_average_vals = []
    
    for k in range(1, threshold):
        k_vals.append(k + 1)
        clusters_k, centroids_k = clustering_method(k + 1, sample, init_centroids[k])
        
        s_av_vals_for_k = []
        
        for cl in clusters_k:
            s_cl = []
            if len(cl) == 1: s_cl.append(0)
            else:
                for i in range(len(cl)):
                    a_i = 0
                    b_i = 0
                    av_dists = []

                    for j in range(len(cl)):
                        if j != i: a_i += metric(cl[i], cl[j])
                    a_i /= (len(cl) - 1)

                    for c in clusters_k:
                        if c == cl: continue
                        else:
                            av_dist = 0
                            for j in range(len(c)):
                                av_dist += metric(cl[i], c[j])
                            av_dist /= len(c)
                            av_dists.append(av_dist)
                    
                    b_i = min(av_dists)
                    s_i = (b_i - a_i) / max(a_i, b_i)
                    s_cl.append(s_i)
            
            s_av_vals_for_k.append(sum(s_cl) / len(s_cl))
        
        s_average_vals.append(sum(s_av_vals_for_k) / len(s_av_vals_for_k))
    
    index_max = max(range(len(s_average_vals)), key=s_average_vals.__getitem__)
    
    plt.grid(True)
    plt.xlabel("Количество кластеров")
    plt.ylabel("Среднее значение коэффициента силуэтов")
    plt.plot(k_vals, s_average_vals, 'r')
    plt.plot(k_vals, s_average_vals, 'ro')
    
    print("Оптимальное количество кластеров: ", index_max + 2)
```

```python
def find_radius_borders(sample):
    dist_mat = []
    for i in range(len(sample)):
        row = []
        for j in range(len(sample)):
            row.append(Euclidian_dist(sample[i], sample[j]))
        dist_mat.append(row)
    
    min_row_vals = []
    max_row_vals = []
    for r in dist_mat:
        min_row_vals.append(min(filter(lambda x:x>0, r)))
        max_row_vals.append(max(r))
    
    R_min = min(min_row_vals)
    R_max = max(max_row_vals)
    
    return R_min, R_max
```

```python
R_min, R_max = find_radius_borders(samp2D_normalized)
print("Диапазон значений радиуса для метода FOREL:")
display(Latex(r'$R_{min} = $' + str(R_min)))
display(Latex(r'$R_{max} = $' + str(R_max)))
```

```python
def FOREL(sample, R):
    clusters = []
    centroids = []
    
    U = copy.deepcopy(sample)
    
    while len(U) > 0:
        old_mean = U[0]
        new_mean = old_mean
        
        K = []
        K_indecies = []
        
        while True:
            old_mean = new_mean
            K = []
            K_indecies = []
            for i in range(len(U)):
                if Euclidian_dist(U[i], old_mean) <= R:
                    K.append(U[i])
                    K_indecies.append(i)
            
            K_x = [k[0] for k in K]
            K_y = [k[1] for k in K]
            
            new_mean = [sum(K_x) / len(K_x), sum(K_y) / len(K_y)]
            
            if old_mean == new_mean: break
        
        centroids.append(old_mean)
        clusters.append(K)
        new_U = [el for i, el in enumerate(U) if i not in K_indecies]
        U = new_U
    
    return clusters, centroids
```

```python
def FOREL_tab(R_vals, sample):
    R_centroids = []
    R_cluster_size = []
    
    for i in range(len(R_vals)):
        cls, cens = FOREL(sample, R_vals[i])
        R_centroids.append(cens)
        
        cls_size = []
        for c in cls: cls_size.append(len(c))
        R_cluster_size.append(cls_size)
    
    df_data = {"$R$": R_vals, "Центроиды $x_{C_i}$": R_centroids, "Объём кластера $C_i$": R_cluster_size}
    df = pd.DataFrame(df_data)
    
    return df
```

<p>В отличие от стандартного K-means, в K-means++ случайным образом выбирается только первый центроид (среди точек выборки). Остальные центроиды (так же среди точек выборки) выбираются так, чтобы они были максимально удалены друг от друга. Это позволяет охватить максимально ввозможную область графика и более равномерно распределить точки по кластерам. Также это может ускорить работу метода K-means.</p>

```python
import sys
```

```python
def get_init_centroids(sample, metric, k):
    centroids = [sample[2]]
    
    for i in range(k - 1):
        dists_and_coords = []
        
        for j in range(len(sample)):
            d = sys.maxsize
            
            for k in range(len(centroids)):
                this_dist = metric(sample[j], centroids[k])
                d = min(this_dist, d)
            dists_and_coords.append([sample[j], d])
        
        dists = [el[1] for el in dists_and_coords]
        index_max = max(range(len(dists)), key=dists.__getitem__)
        
        centroids.append(dists_and_coords[index_max][0])
    
    return centroids
```

```python
def K_means_plus_plus(sample, metric, k):
    init_centroids = get_init_centroids(sample, metric, k)
    
    clusters, centroids = K_means(k, sample, init_centroids)
    
    return clusters, centroids
```
