–†–µ—à–∏–≤ —Å–∏—Å—Ç–µ–º—É, –º—ã –ø–æ–ª—É—á–∞–µ–º –∏—Å–∫–æ–º—ã–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã.

$$\overline{y_x} = r_{yx} \cdot \frac{\sigma_y}{\sigma_x}(x - \overline{x}) + \overline{y}$$
$$\overline{x_y} = r_{xy} \cdot \frac{\sigma_x}{\sigma_y}(y - \overline{y}) + \overline{x},$$

–°—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∞—è –æ—Ü–µ–Ω–∫–∏ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ —á–µ—Ä–µ–∑ —É—Å–ª–æ–≤–Ω—ã–µ –≤–∞—Ä–∏–∞–Ω—Ç—ã: $$r_–≤ = \frac{\frac{1}{n} \cdot \sum{\sum{n_{uv} \cdot u_i \cdot v_j - \overline{u} \cdot \overline{v}}}}{\sigma_u  \cdot \sigma_v}$$

$U_j = \sum{n_{uv} \cdot u_i}$
$V_i = \sum{n_{uv} \cdot v_j}$

–û—Ü–µ–Ω–∫–∞ –∫–∞—á–µ—Å—Ç–≤–∞ –ú–ù–ö - –º–æ–¥–µ–ª–µ–π:
$\eta_{yx} = \frac{\sigma_{\overline{y} x}}{\sigma_y},$

–≥–¥–µ $\sigma_{\overline{y} x} = \sqrt{\frac{\sum{n_x \cdot (\overline{y_x} - \overline{y})^2}}{n}}$ , –∞ $\sigma_y = \sqrt{\frac{\sum{n_y \cdot (y - \overline{y})^2}}{n}}$

–î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª –¥–ª—è –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏:
$z = arth(r_–≤)$, 

—Ç–æ–≥–¥–∞ –∏—Å–∫–æ–º—ã–π –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª –¥–ª—è $z$ —Ä–∞–≤–µ–Ω
$$(z - \frac{t}{\sqrt{n-3}}; z + \frac{t}{\sqrt{n-3}}),$$
–≥–¥–µ $\Phi(t) = \frac{\gamma}{2}$

–¢–æ–≥–¥–∞ –∏—Å–∫–æ—Ä–º—ã–π –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª –¥–ª—è –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ —Ä–∞–≤–µ–Ω:
$$(th z_{left}; th z_{right})$$


–ü—Ä–æ–≤–µ—Ä–∫–∞ –≥–∏–ø–æ—Ç–µ–∑—ã:
$H_0: r_–≤ = 0$
$H_0: r_–≤ \neq 0$
–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ $r_–≤$ —Å–ª–æ–∂–Ω–æ–µ, –ø–æ—ç—Ç–æ–º—É
$$T_{–Ω–∞–±–ª} = r_–≤ \cdot \frac{\sqrt{n-2}}{\sqrt{1-r^2_–≤}}$$

–í—ã–±–æ—Ä–æ—á–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ –µ—Å—Ç—å —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –°—Ç—å—é–¥–µ–Ω—Ç–∞ –ø—Ä–∏ $df = n-2$, –ø–æ—ç—Ç–æ–º—É $T_{–∫—Ä–∏—Ç} = t(\alpha, df)$

–û–±–ª–∞—Å—Ç—å –ø—Ä–∏–Ω—è—Ç–∏—è: $|T_{–Ω–∞–±–ª} < T_{–∫—Ä–∏—Ç}|$


–ü–æ–∫–∞–∑–∞—Ç–µ–ª–∏ –∫–∞—á–µ—Å—Ç–≤–∞ –º–æ–¥–µ–ª–∏:

–¢–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏–π –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –¥–µ—Ç–µ—Ä–º–∏–Ω–∞—Ü–∏–∏ - –æ—Å–Ω–æ–≤–Ω–æ–π –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å, –æ—Ç—Ä–∞–∂–∞—é—â–∏–π –º–µ—Ä—É –∫–∞—á–µ—Å—Ç–≤–∞ —Ä–µ–≥—Ä–µ—Å—Å–∏–æ–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏, –æ–ø–∏—Å—ã–≤–∞—é—â–µ–π —Å–≤—è–∑—å –º–µ–∂–¥—É –∑–∞–≤–∏—Å–∏–º–æ–π –∏ –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã–º–∏ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–º–∏ –º–æ–¥–µ–ª–∏
$$R^2_y = \frac{\frac{1}{n} \cdot \sum{(y_{x_i} - \overline{y})^2}}{\sigma^2_y}$$

–°—Ä–µ–¥–Ω—è—è –∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–∞—è –æ—à–∏–±–∫–∞ - —Å—Ä–µ–¥–Ω–µ–µ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∏–∑ —Å—É–º–º—ã –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ –æ—à–∏–±–æ–∫ –æ—Ç–¥–µ–ª—å–Ω—ã—Ö –∏–∑–º–µ—Ä–µ–Ω–∏–π. –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ –º–æ–¥–µ–ª–∏ —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ —è–≤–ª—è–µ—Ç—Å—è —Ü–µ–ª–µ—Å–æ–æ–±—Ä–∞–∑–Ω—ã–º, –µ—Å–ª–∏ —Å—Ä–µ–¥–Ω—è—è –∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–∞—è –æ—à–∏–±–∫–∞ –º–µ–Ω—å—à–µ, —á–µ–º –°–ö–û
$$S_{\varepsilon y} = \sqrt{\frac{1}{n - l} \cdot \sum{\varepsilon^2_{y_i}}},$$

–≥–¥–µ $l$ - —á–∏—Å–ª–æ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –≤ –º–æ–¥–µ–ª–∏, –∞ $\varepsilon_{y_i} = y_i - y_{x_i}$

–°—Ä–µ–¥–Ω—è—è –æ—à–∏–±–∫–∞ –∞–ø–ø—Ä–æ–∫—Å–∏–º–∞—Ü–∏–∏ –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ —Ä–∞—Å—Å—á–µ—Ç–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –æ—Ç —Ñ–∞–∫—Ç–∏—á–µ—Å–∫–∏—Ö, –µ—Å–ª–∏ –æ–Ω–∞ –º–µ–Ω—å—à–µ 7%, –º–æ–¥–µ–ª—å —Å—á–∏—Ç–∞–µ—Ç—Å—è –º–æ–¥–µ–ª—å—é —Ö–æ—Ä–æ—à–µ–≥–æ –∫–∞—á–µ—Å—Ç–≤–∞
$$A_y = \frac{1}{n} \cdot \sum{|\frac{\varepsilon_y}{y_i}|} \cdot 100% $$

```python
def GetN(start, end, dataset):
    n=0
    for i in dataset:
        if(i>=start and i<=end):
            n+=1
    return n

def ToInterval(dataset):
    k=1+round(np.log2(len(dataset)))
    R=max(dataset)-min(dataset)
    h=R/k
    
    intervalData=np.zeros((k, 7))

    i=0
    n=0
    w=0
    xPrev=min(dataset)
    while xPrev<max(dataset):
        intervalData[i,0]=xPrev #X–Ω–∞—á
        intervalData[i,1]=xPrev+h #X–∫–æ–Ω
        intervalData[i,2]=(2*xPrev+h)/2 #–•—Å—Ä
        intervalData[i,3]=GetN(intervalData[i,0], intervalData[i,1], dataset) #n
        intervalData[i,4]=round(intervalData[i,3]/len(dataset), 3) #p
        n+=intervalData[i, 3]
        intervalData[i,5]=n
        w+=intervalData[i, 4]
        intervalData[i,6]=w
        
        xPrev=xPrev+h
        i+=1
        
    return intervalData, h

def GetData6(dataset, Idataset, h): 
    data6=np.zeros((len(Idataset[:, 0]), 8))
    Cfor6=max(Idataset[round(len(Idataset)/2), 3], Idataset[round(len(Idataset)/2)-1, 3])
    
    index=0
    for i in range(len(Idataset[:, 0])):
        if(Idataset[i, 3]==Cfor6):
            index=i
    
    print("C ", Cfor6, index)

    for i in range(len(Idataset[:, 0])):
        data6[i,0]=Idataset[i, 2] #x
        data6[i,1]=Idataset[i, 3] #n
        data6[i,2]=round((Idataset[i, 2] - Idataset[index, 2])/h, 4) #u
        data6[i,3]=data6[i,1]*data6[i,2] #nu
        data6[i,4]=data6[i,1]*(data6[i,2]**2) #nu^2
        data6[i,5]=data6[i,1]*(data6[i,2]**3) #nu^3
        data6[i,6]=data6[i,1]*(data6[i,2]**4) #nu^4
        data6[i,7]=data6[i,1]*((data6[i,2]+1)**4) #n(u+1)^4
    return data6, Idataset[index, 2]

#data6Print=pd.DataFrame(data6, columns=["X—Å—Ä", "n","u", "nu", "nu^2", "nu^3", "nu^4", "n(u+1)^4"])
#print("–¢–∞–±–ª–∏—Ü–∞ –∑–∞–¥–∞–Ω–∏—è 6 ", "\n", data6Print, "\n")    
#print("–°—É–º–º—ã: \n", "n: ", sum(data6[:,1]), "\n", "nu: ", sum(data6[:,3]), "\n","nu^2: ", sum(data6[:,4]), "\n", "nu^3: ", sum(data6[:,5]), "\n", "nu^4: ", sum(data6[:,6]), "\n", "n(u+1)^4: ", sum(data6[:,7]), "\n")
#print("–ö–æ–Ω—Ç—Ä–æ–ª—å: —Å—É–º–º–∞ n(u+1)^4", sum(data6[:,7]), " = –æ—Å—Ç–∞–ª—å–Ω—ã–º —Å—É–º–º–∞–º", sum(data6[:,6]) + 4*sum(data6[:,5]) + 6*sum(data6[:,4]) + 4*sum(data6[:,3]) + sum( data6[:,1]) )

def GetMoments(data6, dataset, h):
    EmpMoments=np.zeros((4, 2))
    
    for i in range(4):
        j=i+3
        EmpMoments[i,0]=sum(data6[:, j])/len(dataset)
    EmpMoments[0,1]=0
    EmpMoments[1,1]=(EmpMoments[1,0]-(EmpMoments[0,0]**2))*(h**2)    
    EmpMoments[2,1]=(EmpMoments[2,0] - 3*EmpMoments[1,0]*EmpMoments[0,0] + 2*(EmpMoments[0,0]**3))*(h**3)
    EmpMoments[3,1]=(EmpMoments[3,0] - 4*EmpMoments[2,0]*EmpMoments[0,0] + 6*EmpMoments[1,0]*(EmpMoments[0,0]**2) - 3*(EmpMoments[0,0]**4))*(h**4)
    return EmpMoments
    
#EmpMomentsPrint=pd.DataFrame(EmpMoments, columns=["–£—Å–ª–æ–≤–Ω—ã–π —ç.–º.", "–¶–µ–Ω—Ç—Ä–∞–ª—å–Ω—ã–π —ç.–º."])
#print("–¢–∞–±–ª–∏—Ü–∞ —ç–º–ø–∏—Ä–∏—á–µ—Å–∫–∏—Ö –º–æ–º–µ–Ω—Ç–æ–≤: \n", EmpMomentsPrint, "\n")

def ModeMedianVar(Idata5, h, dataset, sko):
    k=0
    max=0
    for i in range(len(Idata5[:, 3])):
        if(Idata5[i, 3]>max):
            max=Idata5[i, 3]
            k=i
            
    print("k", k)
    Nprev=0
    Nnext = 0
    if(k!=0): #–µ—Å–ª–∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π —ç–ª–µ–º–µ–Ω—Ç –∫—Ä–∞–π–Ω–∏–π, —Ç–æ —á–∞—Å—Ç–æ—Ç—É –ø—Ä–µ–¥ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞ –∑–∞–Ω—É–ª—è–µ–º
        Nprev =Idata5[k-1, 3]
    if(k!=len(Idata5[:, 3])-1 ): #–µ—Å–ª–∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π —ç–ª–µ–º–µ–Ω—Ç –∫—Ä–∞–π–Ω–∏–π, —Ç–æ —á–∞—Å—Ç–æ—Ç—É —Å–ª–µ–¥ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞ –∑–∞–Ω—É–ª—è–µ–º
        Nnext =Idata5[k+1, 3]
    
    Mo=Idata5[k, 0] + h*( (Idata5[k, 3]-Nprev) / ( (Idata5[k, 3]-Nprev) + (Idata5[k, 3]-Nnext) ) ) 
    Mo=round(Mo, 2)
    
    k=0
    med=(np.max(dataset)-np.min(dataset))/2
    summP=0
    summ=0
    for i in range(len(Idata5[:, 0])):
        if(summ<0.5 and (summ+Idata5[i, 4])>=0.5):
            k=i
        elif(summ<0.5):
            summP+= Idata5[i, 4]
        summ+=Idata5[i, 4]
        
    summP=round(summP, 4)
            
    
    Me=Idata5[k, 0] + h*( (0.5 - summP )/Idata5[k, 4] )
    Me=round(Me, 2)
    
    V=sko/abs( sum(Idata5[:, 2])/ len(Idata5[:, 2]) )
    V=round(V*100, 2)
    #print("–ú–æ–¥–∞", Mo," –º–µ–¥–∏–∞–Ω–∞ ", Me," –∏ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –≤–∞—Ä–∏–∞—Ü–∏–∏ ", V, "% \n")
    return Mo, Me, V

def GetCharacteristics(Idataset, dataset, EmpMomentsdata, h):
    Characteristics=np.zeros((1,10))
    
    Characteristics[0,0]=round(sum(Idataset[:, 2]*Idataset[:,3])/len(dataset),4)
    Characteristics[0,1]=round(sum(Idataset[:, 3]*(Idataset[:, 2] - Characteristics[0,0])**2)/len(dataset),4)
    
    #EmpExp=h*EmpMoments[0,0]+Cfor6
    #Empdisp=EmpMoments[1,1]
    #print("–°—Ä–∞–≤–Ω–µ–Ω–∏–µ –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏—Ö –æ–∂–∏–¥–∞–Ω–∏–π: ",  Characteristics[0,0] ," = ", EmpExp)
    #print("–°—Ä–∞–≤–Ω–µ–Ω–∏–µ –¥–∏—Å–ø–µ—Ä—Å–∏–π:",  Characteristics[0,1] ," = ", Empdisp)
    
    Characteristics[0,2] = round(np.sqrt(Characteristics[0,1]),4)
    Characteristics[0,3]=round(sum(Idataset[:, 3]*(Idataset[:, 2]-Characteristics[0,0])**2)/(len(dataset)-1),4)
    Characteristics[0,4]=round(np.sqrt(Characteristics[0,3]),4)
    Characteristics[0,5]=round(EmpMomentsdata[2,1]/((Characteristics[0,2])**3),4)
    Characteristics[0,6]=round((EmpMomentsdata[3,1]/((Characteristics[0,2])**4))-3,4)
    Characteristics[0,7], Characteristics[0,8], Characteristics[0,9] = ModeMedianVar(Idataset, h, dataset, Characteristics[0,2])
    Characteristics[0,7] = round(Characteristics[0,7],4)
    Characteristics[0,8] = round(Characteristics[0,8],4)
    Characteristics[0,9] = round(Characteristics[0,9],4)
    return Characteristics

IdataY, hY=ToInterval(data2)
IdataYPrint=pd.DataFrame(IdataY, columns=["Y–Ω–∞—á", "Y–∫–æ–Ω","Y—Å—Ä", "n", "p", "Œ£n", " Œ£p"])
print("–ò–Ω—Ç–µ—Ä–≤–∞–ª—å–Ω—ã–π —Ä—è–¥", "\n", IdataYPrint, "\n")
print("–°—É–º–º—ã ", "\n n:", sum(IdataY[:, 3]), " p:", round(sum(IdataY[:, 4]), 2), "\n")

Ydata6, YCfor6 =GetData6(data2, IdataY, hY)
EmpMomentsY=GetMoments(Ydata6, data2, hY)

data2Characteristics = GetCharacteristics(IdataY, data2, EmpMomentsY, hY)
np.set_printoptions(suppress=True)
data2CharacteristicsPrint=pd.DataFrame(data2Characteristics, columns=["Y–≤—ã–±", "œÉ^2", "œÉ", "s^2", "s", "a*", "Œµ*", "Mo*", "Me*", "V*"])
print("–¢–∞–±–ª–∏—Ü–∞ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫ Y: \n", data2CharacteristicsPrint, "\n")


IdataX, hX=ToInterval(data1)
IdataXPrint=pd.DataFrame(IdataX, columns=["X–Ω–∞—á", "X–∫–æ–Ω","X—Å—Ä", "n", "p", "Œ£n", " Œ£p"])
print("–ò–Ω—Ç–µ—Ä–≤–∞–ª—å–Ω—ã–π —Ä—è–¥", "\n", IdataXPrint, "\n")
print("–°—É–º–º—ã ", "\n n:", sum(IdataX[:, 3]), " p:", round(sum(IdataX[:, 4]), 2), "\n")

Xdata6, XCfor6 = GetData6(data1, IdataX, hX)
EmpMomentsX = GetMoments(Xdata6, data1, hX)

data1Characteristics = GetCharacteristics(IdataX, data1, EmpMomentsX, hX)
np.set_printoptions(suppress=True)
data1CharacteristicsPrint=pd.DataFrame(data1Characteristics, columns=["X–≤—ã–±", "œÉ^2", "œÉ", "s^2", "s", "a*", "Œµ*", "Mo*", "Me*", "V*"])
print("–¢–∞–±–ª–∏—Ü–∞ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫ X: \n", data1CharacteristicsPrint, "\n")
```

```python
def KoeffKorrelation(TableUVData, uDataset, vDataset, YCharacteristics, XCharacteristics):
    
    
    summ = 0
    for i in range(len(uDataset)):
        summ += TableUVData[8,i]*uDataset[i]
    expU = round(summ/len(uDataset),4)
    
    summ = 0
    for i in range(len(vDataset)):
        summ += TableUVData[i,8]*vDataset[i]
    expV = round(summ/len(vDataset),4)
    #print("–º–∞—Ç –æ–∂–∏–¥–∞–Ω–∏—è u –∏ v ", expU, expV)
    
    summ = 0
    for i in range(len(uDataset)):
        summ += TableUVData[8,i]*(uDataset[i]-expU)**2
    SKOU = round(np.sqrt(summ/len(uDataset)),4)
    
    summ = 0
    for i in range(len(vDataset)):
        summ += TableUVData[i,8]*(vDataset[i] - expV)**2
    SKOV = round(np.sqrt(summ/len(vDataset)),4)
    #print("–°–ö–û u –∏ v ", SKOU, SKOV)
    
    summ = 0
    for i in range(len(vDataset)):
        for j in range(len(uDataset)):
            summ+= (TableUVData[j, i]*uDataset[j]*vDataset[i]) - (expU*expV)
    KoeffKorr =  summ/(TableUVData[8,8]*SKOU*SKOV)
    print("–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏: ", KoeffKorr)
    
    summ = 0
    for i in range(len(vDataset)):
        summ+=TableUVData[10,i] -(expU*expV) #vDataset[i]*TableUVData[9,i]
    KoeffKorrU = summ/(TableUVData[8,8]*SKOU*SKOV)
    print("–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ —á–µ—Ä–µ–∑ U: ", KoeffKorr)

    summ = 0
    for i in range(len(uDataset)):
        summ+=TableUVData[i,10] -(expU*expV) #uDataset[i]*TableUVData[i,9]
    KoeffKorrV = summ/(TableUVData[8,8]*SKOU*SKOV)
    print("–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ —á–µ—Ä–µ–∑ V: ", KoeffKorr)
    
    return KoeffKorr


KoeffKorr = KoeffKorrelation(TableUV, ui, vi, data2Characteristics, data1Characteristics)
```

```python

def GetTrustInterval(KoeffKorrData, TableUVData):
    
    #–§—É–Ω–∫—Ü–∏—è –ª–∞–ø–ª–∞—Å–∞(t) = 0,475 –∏ 0,495 –ø—Ä–∏ t = 1,96 –∏ t = 2.58
    t95=1.96
    t99=2.58
    
    z = math.atanh(KoeffKorrData)
    left95=z - ( t95/np.sqrt(TableUVData[8,8]-3) )
    right95=z + ( t95/np.sqrt(TableUVData[8,8]-3) )
    left99=z - ( t99/np.sqrt(TableUVData[8,8]-3) )
    right99=z + ( t99/np.sqrt(TableUVData[8,8]-3) )
    
    left95 = math.tan(left95)
    right95 = math.tan(right95)
    left99 = math.tan(left99)
    right99 = math.tan(right99)
    print("–î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª –∫–æ—ç—Ñ—Ñ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –¥–ª—è 95%: (", left95, " ; ", right95, ")")
    print("–î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª –∫–æ—ç—Ñ—Ñ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –¥–ª—è 99%: (", left99, " ; ", right99, ")")
    
GetTrustInterval(KoeffKorr, TableUV)
```
