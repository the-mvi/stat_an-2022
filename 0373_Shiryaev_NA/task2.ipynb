## Основные теоретические положения


Интервальный вариационный ряд - ряд, в котором промежуток выборки разбивается на частичные интервалы. Интервальный вариационный ряд формируется по формуле Стёрджерса: $k=1+\log_2(n),$ размаху выборки $X_{max}-X_{min}$ и длине $h=\frac{R}{k}$

Выборочное среднее - среднее арифметических значений выборочной совокупности $x_в = \frac{1}{n} \cdot \sum{x_i*n_i}$

Выборочная дисперсия - мера разброса значений случайной величины относительно её математического ожидания 
$\sigma^2 = \frac{1}{n} \cdot \sum{n_i \cdot (x_i-x_в)^2}$.

Мода интервального ряда - число в частичном интервале с наибольшей частотой.
$$M_o = x_{M_o}^{(0)}+h \cdot \frac{n_{M_o}-n_{M_o-1}}{(n_{M_o}-n_{M_o-1})+(n_{M_o}-n_{M_o+1})},$$
где $x_{M_o}^{(0)}$ - левый край интервала с наибольшей частотой, $n_{M_o}$ - частота модального интервала, $n_{M_o+1} и n_{M_o-1}$ - частоты следущего и предыдущего интервалов соответственно. Если интервал с наибольшей частотой крайний, то частоту 
предыдущего интервала(если он первый) или последующего(если он последний) зануляем.

Медиана интервального ряда - число интервала, для которого частота составляет половину или больше половины всей суммы частот, а предыдущая накопленная частота меньше половины всей суммы частот. 
$$M_e = x_{M_e}^{(0)}+h \cdot \frac{0.5-p_{M_e-1}^{*}}{p_{M_e}^{*}},$$
где $x_{M_е}^{(0)}$ - левый край интервала с наибольшей частотой, $p_{M_e-1}^{*}$ - относительная частота медианного интервала, $n_{M_o-1}$ - относительные частоты последующего интервалов соответственно.

Коэффициент вариации - мера колеблемости признака относительно выборочной средней. $V^* = \frac{\sigma_в}{|x_в|}*100\%$

Условные варианты - варианты, определяемые по неравенству:
$$u_i=\frac{x_i-C}{h},$$ где С - ложный нуль(варианта с наибольшей частотой в середине). 

Условные эмпирический момент порядка r - начальный эмпирический момент порядка r, вычисленный с помощью условных вариант. 
$$\nu_r^* = \frac{1}{n} \cdot \sum{n_i \cdot u_i^r},$$ причем $x_в = h \cdot \nu_1^* + C$
$$\mu_1^* = 0$$
$$\mu_2^* = h^2 \cdot (\nu_2^* - (\nu_1^*)^2)$$
$$\mu_3^* = h^3 \cdot (\nu_3^* - 3 \cdot \nu_2^*\nu_1^* + 2 \cdot (\nu_1^*)^3)$$
$$\mu_4^* = h^4 \cdot (\nu_4^* - 4 \cdot \nu_3^*\nu_1^* + 6 \cdot \nu_2^* \cdot (\nu_1^*)^2 - 3 \cdot (\nu_1^*)^4)$$

Выборочный коэффициент ассиметрии: $a_s^* = \frac{\mu_3^*}{\sigma_в^3}$

Выборочный коэффициент эксцесса: $\varepsilon_k^* = \frac{\mu_4^*}{\sigma_в^4} - 3$

Исправленная дисперсия: $s^2 = \frac{1}{n-1} \cdot \sum{n_i*(x_i - x_в)^2}$


Статистическая гипотеза - любое предположение о генеральной совокупности, проверяемое по выборке.

Статистический критерий проверки основной гипотезы - правило, по которому принимается решение принять или отклонить гипотезу.

Статистика критерия - случайная величина, служащая для проверки относительных гипотез.

Критическая область - совокупность значений критерия, при которых основную гипотезу отвергают.


Функциональная зависимость - каждому значению одной переменной соответствует единственное значение другой пременной.

Статистическая зависимость - изменение одной из величин влечет изменение распределения другой.

Корреляционная зависимость - изменение одной величины приводит к изменению среднего значения другой.

Условное среднее - среднее арифметическое тех значений случайных величин $X(Y)$, которые соотвестсвуют значению случайной величины $Y = y(X = x)$:
$$x_{y} = \frac{\sum{x_i \cdot n_{x_i y_i}}}{\sum{n_{x_i y_i}}}$$

Межгрупповое среднее: $\overline{xy} = \frac{1}{n} \cdot \sum{\sum{n_{x_i y_j} \cdot x_i \cdot y_j}}$

Уравнение регрессии: $E(Y|x) = f(x)$  $E(X|y) = f(y)$

Выборочные уравнения регрессии: 
$$\overline{y_x} = f^*(x)$$  
$$\overline{x_y} = f^*(y)$$

Метод наименьших квадратов:
основан на минимизации сумм квадратов отклонений функции от искромых переменных. 
$$F(a_1 \cdots a_m) = \sum{(y_i - f(x_i; a_1 \cdots a_m))^2} -> min$$

Далее находим частные производные функции и приравниваем их к нулю:

$$\begin{equation*}
 \begin{cases}
   \frac{dF}{d a_1} = 0
   \\
   \cdots
   \\
   \frac{dF}{d a_m} = 0
 \end{cases}
\end{equation*}$$

Решив систему, мы получаем искомые коэффициенты.

Уравнения средней квадратичной регрессии:
$$\overline{y_x} = r_{yx} \cdot \frac{\sigma_y}{\sigma_x}(x - \overline{x}) + \overline{y}$$
$$\overline{x_y} = r_{xy} \cdot \frac{\sigma_x}{\sigma_y}(y - \overline{y}) + \overline{x},$$ 
где $r_{yx} = r_{xy}$ - статистическая оценки коэффициента корреляции $r_в = \frac{\mu_{xy}}{s_x \cdot s_y} = \frac{\overline{xy} - \overline{x} \cdot \overline{y}}{s_x \cdot s_y}$

Статистическая оценки коэффициента корреляции через условные варианты: $$r_в = \frac{\frac{1}{n} \cdot \sum{\sum{n_{uv} \cdot u_i \cdot v_j - \overline{u} \cdot \overline{v}}}}{\sigma_u  \cdot \sigma_v}$$

$U_j = \sum{n_{uv} \cdot u_i}$
$V_i = \sum{n_{uv} \cdot v_j}$

Оценка качества МНК - моделей:
$\eta_{yx} = \frac{\sigma_{\overline{y} x}}{\sigma_y},$

где $\sigma_{\overline{y} x} = \sqrt{\frac{\sum{n_x \cdot (\overline{y_x} - \overline{y})^2}}{n}}$ , а $\sigma_y = \sqrt{\frac{\sum{n_y \cdot (y - \overline{y})^2}}{n}}$

Доверительный интервал для коэффициента корреляции:
$z = arth(r_в)$, 

тогда искомый доверительный интервал для $z$ равен
$$(z - \frac{t}{\sqrt{n-3}}; z + \frac{t}{\sqrt{n-3}}),$$
где $\Phi(t) = \frac{\gamma}{2}$

Тогда искормый доверительный интервал для коэффициента корреляции равен:
$$(th z_{left}; th z_{right})$$


Проверка гипотезы:
$H_0: r_в = 0$
$H_0: r_в \neq 0$
Распределение $r_в$ сложное, поэтому
$$T_{набл} = r_в \cdot \frac{\sqrt{n-2}}{\sqrt{1-r^2_в}}$$

Выборочное распределение статистики есть распределение Стьюдента при $df = n-2$, поэтому $T_{крит} = t(\alpha, df)$

Область принятия: $|T_{набл} < T_{крит}|$


Показатели качества модели:

Теоретический коэффициент детерминации - основной показатель, отражающий меру качества регрессионной модели, описывающей связь между зависимой и независимыми переменными модели
$$R^2_y = \frac{\frac{1}{n} \cdot \sum{(y_{x_i} - \overline{y})^2}}{\sigma^2_y}$$

Средняя квадратичная ошибка - среднее квадратичное значение из суммы квадратов ошибок отдельных измерений. Использование модели регрессии является целесообразным, если средняя квадратичная ошибка меньше, чем СКО
$$S_{\varepsilon y} = \sqrt{\frac{1}{n - l} \cdot \sum{\varepsilon^2_{y_i}}},$$

где $l$ - число параметров в модели, а $\varepsilon_{y_i} = y_i - y_{x_i}$

Средняя ошибка аппроксимации показывает отклонение рассчетных значений от фактических, если она меньше 7%, модель считается моделью хорошего качества
$$A_y = \frac{1}{n} \cdot \sum{|\frac{\varepsilon_y}{y_i}|} \cdot 100% $$

```python
def GetN(start, end, dataset):
    n=0
    for i in dataset:
        if(i>=start and i<=end):
            n+=1
    return n

def ToInterval(dataset):
    k=1+round(np.log2(len(dataset)))
    R=max(dataset)-min(dataset)
    h=R/k
    
    intervalData=np.zeros((k, 7))

    i=0
    n=0
    w=0
    xPrev=min(dataset)
    while xPrev<max(dataset):
        intervalData[i,0]=xPrev #Xнач
        intervalData[i,1]=xPrev+h #Xкон
        intervalData[i,2]=(2*xPrev+h)/2 #Хср
        intervalData[i,3]=GetN(intervalData[i,0], intervalData[i,1], dataset) #n
        intervalData[i,4]=round(intervalData[i,3]/len(dataset), 3) #p
        n+=intervalData[i, 3]
        intervalData[i,5]=n
        w+=intervalData[i, 4]
        intervalData[i,6]=w
        
        xPrev=xPrev+h
        i+=1
        
    return intervalData, h

def GetData6(dataset, Idataset, h): 
    data6=np.zeros((len(Idataset[:, 0]), 8))
    Cfor6=max(Idataset[round(len(Idataset)/2), 3], Idataset[round(len(Idataset)/2)-1, 3])
    
    index=0
    for i in range(len(Idataset[:, 0])):
        if(Idataset[i, 3]==Cfor6):
            index=i
    
    print("C ", Cfor6, index)

    for i in range(len(Idataset[:, 0])):
        data6[i,0]=Idataset[i, 2] #x
        data6[i,1]=Idataset[i, 3] #n
        data6[i,2]=round((Idataset[i, 2] - Idataset[index, 2])/h, 4) #u
        data6[i,3]=data6[i,1]*data6[i,2] #nu
        data6[i,4]=data6[i,1]*(data6[i,2]**2) #nu^2
        data6[i,5]=data6[i,1]*(data6[i,2]**3) #nu^3
        data6[i,6]=data6[i,1]*(data6[i,2]**4) #nu^4
        data6[i,7]=data6[i,1]*((data6[i,2]+1)**4) #n(u+1)^4
    return data6, Idataset[index, 2]

#data6Print=pd.DataFrame(data6, columns=["Xср", "n","u", "nu", "nu^2", "nu^3", "nu^4", "n(u+1)^4"])
#print("Таблица задания 6 ", "\n", data6Print, "\n")    
#print("Суммы: \n", "n: ", sum(data6[:,1]), "\n", "nu: ", sum(data6[:,3]), "\n","nu^2: ", sum(data6[:,4]), "\n", "nu^3: ", sum(data6[:,5]), "\n", "nu^4: ", sum(data6[:,6]), "\n", "n(u+1)^4: ", sum(data6[:,7]), "\n")
#print("Контроль: сумма n(u+1)^4", sum(data6[:,7]), " = остальным суммам", sum(data6[:,6]) + 4*sum(data6[:,5]) + 6*sum(data6[:,4]) + 4*sum(data6[:,3]) + sum( data6[:,1]) )

def GetMoments(data6, dataset, h):
    EmpMoments=np.zeros((4, 2))
    
    for i in range(4):
        j=i+3
        EmpMoments[i,0]=sum(data6[:, j])/len(dataset)
    EmpMoments[0,1]=0
    EmpMoments[1,1]=(EmpMoments[1,0]-(EmpMoments[0,0]**2))*(h**2)    
    EmpMoments[2,1]=(EmpMoments[2,0] - 3*EmpMoments[1,0]*EmpMoments[0,0] + 2*(EmpMoments[0,0]**3))*(h**3)
    EmpMoments[3,1]=(EmpMoments[3,0] - 4*EmpMoments[2,0]*EmpMoments[0,0] + 6*EmpMoments[1,0]*(EmpMoments[0,0]**2) - 3*(EmpMoments[0,0]**4))*(h**4)
    return EmpMoments
    
#EmpMomentsPrint=pd.DataFrame(EmpMoments, columns=["Условный э.м.", "Центральный э.м."])
#print("Таблица эмпирических моментов: \n", EmpMomentsPrint, "\n")

def ModeMedianVar(Idata5, h, dataset, sko):
    k=0
    max=0
    for i in range(len(Idata5[:, 3])):
        if(Idata5[i, 3]>max):
            max=Idata5[i, 3]
            k=i
            
    print("k", k)
    Nprev=0
    Nnext = 0
    if(k!=0): #если максимальный элемент крайний, то частоту пред интервала зануляем
        Nprev =Idata5[k-1, 3]
    if(k!=len(Idata5[:, 3])-1 ): #если максимальный элемент крайний, то частоту след интервала зануляем
        Nnext =Idata5[k+1, 3]
    
    Mo=Idata5[k, 0] + h*( (Idata5[k, 3]-Nprev) / ( (Idata5[k, 3]-Nprev) + (Idata5[k, 3]-Nnext) ) ) 
    Mo=round(Mo, 2)
    
    k=0
    med=(np.max(dataset)-np.min(dataset))/2
    summP=0
    summ=0
    for i in range(len(Idata5[:, 0])):
        if(summ<0.5 and (summ+Idata5[i, 4])>=0.5):
            k=i
        elif(summ<0.5):
            summP+= Idata5[i, 4]
        summ+=Idata5[i, 4]
        
    summP=round(summP, 4)
            
    
    Me=Idata5[k, 0] + h*( (0.5 - summP )/Idata5[k, 4] )
    Me=round(Me, 2)
    
    V=sko/abs( sum(Idata5[:, 2])/ len(Idata5[:, 2]) )
    V=round(V*100, 2)
    #print("Мода", Mo," медиана ", Me," и коэффициент вариации ", V, "% \n")
    return Mo, Me, V

def GetCharacteristics(Idataset, dataset, EmpMomentsdata, h):
    Characteristics=np.zeros((1,10))
    
    Characteristics[0,0]=round(sum(Idataset[:, 2]*Idataset[:,3])/len(dataset),4)
    Characteristics[0,1]=round(sum(Idataset[:, 3]*(Idataset[:, 2] - Characteristics[0,0])**2)/len(dataset),4)
    
    #EmpExp=h*EmpMoments[0,0]+Cfor6
    #Empdisp=EmpMoments[1,1]
    #print("Сравнение математических ожиданий: ",  Characteristics[0,0] ," = ", EmpExp)
    #print("Сравнение дисперсий:",  Characteristics[0,1] ," = ", Empdisp)
    
    Characteristics[0,2] = round(np.sqrt(Characteristics[0,1]),4)
    Characteristics[0,3]=round(sum(Idataset[:, 3]*(Idataset[:, 2]-Characteristics[0,0])**2)/(len(dataset)-1),4)
    Characteristics[0,4]=round(np.sqrt(Characteristics[0,3]),4)
    Characteristics[0,5]=round(EmpMomentsdata[2,1]/((Characteristics[0,2])**3),4)
    Characteristics[0,6]=round((EmpMomentsdata[3,1]/((Characteristics[0,2])**4))-3,4)
    Characteristics[0,7], Characteristics[0,8], Characteristics[0,9] = ModeMedianVar(Idataset, h, dataset, Characteristics[0,2])
    Characteristics[0,7] = round(Characteristics[0,7],4)
    Characteristics[0,8] = round(Characteristics[0,8],4)
    Characteristics[0,9] = round(Characteristics[0,9],4)
    return Characteristics

IdataY, hY=ToInterval(data2)
IdataYPrint=pd.DataFrame(IdataY, columns=["Yнач", "Yкон","Yср", "n", "p", "Σn", " Σp"])
print("Интервальный ряд", "\n", IdataYPrint, "\n")
print("Суммы ", "\n n:", sum(IdataY[:, 3]), " p:", round(sum(IdataY[:, 4]), 2), "\n")

Ydata6, YCfor6 =GetData6(data2, IdataY, hY)
EmpMomentsY=GetMoments(Ydata6, data2, hY)

data2Characteristics = GetCharacteristics(IdataY, data2, EmpMomentsY, hY)
np.set_printoptions(suppress=True)
data2CharacteristicsPrint=pd.DataFrame(data2Characteristics, columns=["Yвыб", "σ^2", "σ", "s^2", "s", "a*", "ε*", "Mo*", "Me*", "V*"])
print("Таблица характеристик Y: \n", data2CharacteristicsPrint, "\n")


IdataX, hX=ToInterval(data1)
IdataXPrint=pd.DataFrame(IdataX, columns=["Xнач", "Xкон","Xср", "n", "p", "Σn", " Σp"])
print("Интервальный ряд", "\n", IdataXPrint, "\n")
print("Суммы ", "\n n:", sum(IdataX[:, 3]), " p:", round(sum(IdataX[:, 4]), 2), "\n")

Xdata6, XCfor6 = GetData6(data1, IdataX, hX)
EmpMomentsX = GetMoments(Xdata6, data1, hX)

data1Characteristics = GetCharacteristics(IdataX, data1, EmpMomentsX, hX)
np.set_printoptions(suppress=True)
data1CharacteristicsPrint=pd.DataFrame(data1Characteristics, columns=["Xвыб", "σ^2", "σ", "s^2", "s", "a*", "ε*", "Mo*", "Me*", "V*"])
print("Таблица характеристик X: \n", data1CharacteristicsPrint, "\n")
```

```python
def GetUVTable(ToTwoDimIntervalDataset, uDataset, vDataset):
    
    TableUV = np.zeros((len(ToTwoDimIntervalDataset[0,:])+2, len(ToTwoDimIntervalDataset[:,0])+2))
    
    for i in range(len(ToTwoDimIntervalDataset[:, 0])):
        TableUV[i, 0] = ToTwoDimIntervalDataset[i,0]
        TableUV[i, 1] = ToTwoDimIntervalDataset[i,1]
        TableUV[i, 2] = ToTwoDimIntervalDataset[i,2]
        TableUV[i, 3] = ToTwoDimIntervalDataset[i,3]
        TableUV[i, 4] = ToTwoDimIntervalDataset[i,4]
        TableUV[i, 5] = ToTwoDimIntervalDataset[i,5]
        TableUV[i, 6] = ToTwoDimIntervalDataset[i,6]
        TableUV[i, 7] = ToTwoDimIntervalDataset[i,7]
        TableUV[i, 8] = ToTwoDimIntervalDataset[i,8] 
    
    for i in range(len(vDataset)): #Ui
        for j in range(len(uDataset)):
            TableUV[i,9]+=TableUV[i,j]*uDataset[j] 

    for i in range(len(uDataset)): #Vi
        for j in range(len(vDataset)):
            TableUV[9,i]+=TableUV[j,i]*vDataset[j]
            
    for i in range(len(vDataset)):#vU
        TableUV[i,10] = vDataset[i]*TableUV[i,9]
        
    for i in range(len(uDataset)):#uV
        TableUV[10,i] = uDataset[i]*TableUV[9,i]
    
    TableUV[10,10] = sum(TableUV[10,:])
        
    TableUVPrint=pd.DataFrame(TableUV, 
                              index = ["-3", "-2", "-1", "0", "1", "2", "3", "4", "nu", "Vi", "uV"], 
                              columns = ["-3", "-2", "-1", "0", "1", "2", "3", "4", "nv", "Uj", "vU"] )
    print("Корреляционная таблица условных вариант: \n", TableUVPrint, "\n")   
    
    return TableUV
    
    
#Ydata6Print=pd.DataFrame(Ydata6, columns=["Yср", "n","u", "nu", "nu^2", "nu^3", "nu^4", "n(u+1)^4"])
#print("Таблица Y задания 6", Ydata6)
#Xdata6Print=pd.DataFrame(Xdata6, columns=["Xср", "n","u", "nu", "nu^2", "nu^3", "nu^4", "n(u+1)^4"])
#print("Таблица X задания 6", Xdata6)

ui = [x for x in Xdata6[:,2]]
vi = [y for y in Ydata6[:,2]]
#print(ui, vi)
TableUV = GetUVTable(ToTwoDimIntervalData, ui, vi)
```

```python
def KoeffKorrelation(TableUVData, uDataset, vDataset, YCharacteristics, XCharacteristics):
    
    
    summ = 0
    for i in range(len(uDataset)):
        summ += TableUVData[8,i]*uDataset[i]
    expU = round(summ/len(uDataset),4)
    
    summ = 0
    for i in range(len(vDataset)):
        summ += TableUVData[i,8]*vDataset[i]
    expV = round(summ/len(vDataset),4)
    #print("мат ожидания u и v ", expU, expV)
    
    summ = 0
    for i in range(len(uDataset)):
        summ += TableUVData[8,i]*(uDataset[i]-expU)**2
    SKOU = round(np.sqrt(summ/len(uDataset)),4)
    
    summ = 0
    for i in range(len(vDataset)):
        summ += TableUVData[i,8]*(vDataset[i] - expV)**2
    SKOV = round(np.sqrt(summ/len(vDataset)),4)
    #print("СКО u и v ", SKOU, SKOV)
    
    summ = 0
    for i in range(len(vDataset)):
        for j in range(len(uDataset)):
            summ+= (TableUVData[j, i]*uDataset[j]*vDataset[i]) - (expU*expV)
    KoeffKorr =  summ/(TableUVData[8,8]*SKOU*SKOV)
    print("Коэффициент корреляции: ", KoeffKorr)
    
    summ = 0
    for i in range(len(vDataset)):
        summ+=TableUVData[10,i] -(expU*expV) #vDataset[i]*TableUVData[9,i]
    KoeffKorrU = summ/(TableUVData[8,8]*SKOU*SKOV)
    print("Коэффициент корреляции через U: ", KoeffKorr)

    summ = 0
    for i in range(len(uDataset)):
        summ+=TableUVData[i,10] -(expU*expV) #uDataset[i]*TableUVData[i,9]
    KoeffKorrV = summ/(TableUVData[8,8]*SKOU*SKOV)
    print("Коэффициент корреляции через V: ", KoeffKorr)
    
    return KoeffKorr


KoeffKorr = KoeffKorrelation(TableUV, ui, vi, data2Characteristics, data1Characteristics)
```

Констрольные суммы равны, сама корреляция находится на промежутке [0; 1]. Близкая к 0 в нашем случае корреляция показывает отсуствие линейной связи.

```python

def GetTrustInterval(KoeffKorrData, TableUVData):
    
    #Функция лапласа(t) = 0,475 и 0,495 при t = 1,96 и t = 2.58
    t95=1.96
    t99=2.58
    
    z = math.atanh(KoeffKorrData)
    left95=z - ( t95/np.sqrt(TableUVData[8,8]-3) )
    right95=z + ( t95/np.sqrt(TableUVData[8,8]-3) )
    left99=z - ( t99/np.sqrt(TableUVData[8,8]-3) )
    right99=z + ( t99/np.sqrt(TableUVData[8,8]-3) )
    
    left95 = math.tan(left95)
    right95 = math.tan(right95)
    left99 = math.tan(left99)
    right99 = math.tan(right99)
    print("Доверительный интервал коэфф корреляции для 95%: (", left95, " ; ", right95, ")")
    print("Доверительный интервал коэфф корреляции для 99%: (", left99, " ; ", right99, ")")
    
GetTrustInterval(KoeffKorr, TableUV)
```

```python
def KorrParabola(massX, massY, length):
    
    a,b,c,x,y = syp.symbols('a b c x y')
    print("Решаем систему производных относительно нуля:") 
    print(syp.diff((a*x**2+b*x+c-y)**2, a))
    print(syp.diff((a*x**2+b*x+c-y)**2, b))
    print(syp.diff((a*x**2+b*x+c-y)**2, c))
    
    massX2 = massX**2
    massX3 = massX**3
    massX4 = massX**4
    massYX = massY*massX
    massYX2 = massY*massX2
    
    left = np.array([[length, sum(massX), sum(massX2)], 
                     [sum(massX), sum(massX2), sum(massX3)], 
                     [sum(massX2), sum(massX3), sum(massX4)]])
    
    right = np.array([sum(massY), sum(massYX), sum(massYX2)])
    koeff = np.linalg.inv(left).dot(right)
    print("Коэффициенты c b a: ", koeff)
    
    Ycoord =[]
    for x in massX:
        Ycoord.append(koeff[2]*x**2 + koeff[1]*x + koeff[0])
    
    #Ycoord.sort()
    
    fig, ax = plt.subplots()
    ax.scatter(csvX, csvY, c = 'b')
    ax.plot(csvX, Ycoord, c = 'r')
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.show()
    
    return koeff[2], koeff[1], koeff[0]
    

A,B,C = KorrParabola(csvX, csvY, ToTwoDimIntervalData[8,8])
```

При помощи их был высчитан коэффициент ковариации, который был равен 0.070324. Правильность вычислений была подтверждена контрольными суммами, а также попаданием в доверительный интервал при уровне значимости 95% и 99% и подтверждением гипотезы о равенстве коэффициента нулю при уровне значимости $\alpha = 0.05$. 

Были построены и выведены графики регрессионных уравнений:
1. Уравнения средней квадратичной регрессии $x$ на $y$ и $y$ на $x$ соответственно, которые в ходе анализа показателей качества показали средние значения вполне целесообразной модели.

2. Корреляционная кривая параболического вида $a\cdot X^2 + b\cdot X + c$, которое в ходе анализа показателей качества вывели лучшие данные, все пороги не были превышены. Модель определенно целесообразна.

3. Прямая логарифмического вида превысила все пороги целесообразности в 7% средней ошибки апроксимации и порога для средней квадратичной ошибки в виде неисправленное ско.
