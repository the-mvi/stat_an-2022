## Основные теоретические положения  
  
#### Для построения интервального ряда используются следующие формулы:  

По формуле Старджесса находим оптимальное количество интервалов: $ k = 1 + \log_2 n $  
Длина каждого интервала равна: $ h = \frac{x_{max} - x_{min}}{k} $  
$ x_{0} = x_{min} $  
$ x_{i} = x_{0} + ih $  

#### Формула середины интервалов:
$ \tilde{x} = \frac{x_{i - 1} + x_{i}}{2} $  

#### Формула относительной частоты:
$ p^{*}_{i} = \frac{n_{i}}{n} $  
  
#### Формула условной варианты:
  
$ u_{i} = \frac{x_{i} - C}{h} $, где $ C $ - ложный нуль, $ h $ - шаг  
Ложный нуль выбирается как варианта, находящаяся посередине. Если количество интервалов чётное, то выбирается варианта с наибольшей частотой.  
  
#### Формула условного эмпирического момента:  
$ \nu^*_{r} = \frac{1}{n}\sum^k_{i = 1}n_iu^r_i $  

#### Формулы первых четырёх центральных эмпирических моментов:  
$ \mu^*_{1} = 0 $  
$ \mu^*_{2} = (\nu^*_{2} - (\nu^*_{1})^2)h^2 $  
$ \mu^*_{3} = (\nu^*_{3} - 3\nu^*_{1}\nu^*_{2} + 2(\nu^*_{1})^3)h^3 $  
$ \mu^*_{4} = (\nu^*_{4} - 4\nu^*_{1}\nu^*_{3} + 6\nu^*_{2}(\nu^*_{1})^2 - 3(\nu^*_{1})^4)h^4 $  
  
#### Формула выборочного среднего:  
$ \bar{x}_{B} = \frac{1}{n}\sum^n_{i + 1}x_in_i $  
$ \bar{x}_{B} = h\nu^*_1 + C $  
  
#### Формула выборочной дисперсии:  
$ \sigma^2_{B} = \frac{1}{n}\sum^n_{i = 1}(x_i - \bar{x}_B)^2 $  
$ \sigma^2_{B} = \nu^*_2 = \frac{1}{n}\sum^k_{i = 1}n_iu^2_i $  

#### Формула исправленных выборочных дисперсии и СКО:
$ s^2 = \frac{n}{n - 1}\sigma^2_B = \frac{1}{n - 1}\sum^k_{i = 1}n_i(x_i - \bar{x_B})^2 $  
$ s = \sqrt{s^2} $  
  
#### Формулы коэффициентов асимметрии и эксцесса:
$ a^*_s = \frac{\mu^*_3}{\sigma^3_B} $  
$ \varepsilon^*_k = \frac{\mu^*_4}{\sigma^4_B} - 3 $  
  
#### Формула моды:  
$ M^*_o \in (x^o_{M_o}, x^{'}_{M_o}); M^*_o = x^o_{M_o} + h\frac{n_{M_o} - n_{M_o - 1}}{(n_{M_o} - n_{M_o - 1}) + (n_{M_o} - n_{M_o + 1})} $  

#### Формула медианы:   
$ M^*_e \in (x^o_{M_e}, x^{'}_{M_e}); M^*_e = x^o_{M_e} + \frac{h}{p^*_{M_e}}(0.5 - p^{\sum}_{M_e - 1}) $  
  
#### Формула коэффициента вариации:   
$ \upsilon^* = \frac{\sigma_B}{|\bar{X_B}|}*100  $  
  
#### Формула доверительного интервала для случая, когда неизвестно СКО:
$ (\bar{x_B} - \frac{t_\gamma s}{\sqrt{n}}, \bar{x_B} + \frac{t_\gamma s}{\sqrt{n}}) $  
  
#### Формула доверительного интервала для СКО:
$ (s(1 - q), s(1 + q)) $  
  
#### Формула доверительного интервала для $ \chi $:  
$ \left ( \frac{\sqrt{n - 1}}{1 + q} < \chi < \frac{\sqrt{n - 1}}{1 - q} \right ) $  

#### Формула точной вероятности $ P_i $:  
$ P_i(x_{i - 1} < x < x_i) = \Phi(\frac{x_i - a}{\sigma }) - \Phi(\frac{x_{i - 1} - a}{\sigma }) $  

#### Формула для $ \chi_{nabl} $:  
$ \chi^2_{nabl} = \sum^k_{i = 1}\frac{n^2_i}{n^{'}_i} - n $, где $ n^{'}_i = nP_i $  
  
#### Формула для $ \chi_{krit} $:  
$ \chi^2_{krit} = \chi^2(\alpha, df) $, где $ df = k - r - 1 $  

#### Условие принятия гипотезы:  
$ \chi^2_{krit} > \chi^2_{nabl}  $

```python
import numpy as np
import pandas as pd
import seaborn as sns
from tabulate import tabulate
import matplotlib.pyplot as plt
import math
import scipy.stats
```

```python
# 1.

print("\nИспользовался серийный способ, т.к. единицы генеральной совокупности изначально объединены в небольшие более или менее равновеликие серии.")
print("\nВыборка дискретная.")

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# Вся выборка
full_sample = pd.read_csv('sample_task1.csv', names=['average_price', 'total_volume'])

# Одномерная выборка
sample = full_sample[['average_price']]

sample
```

```python
# 2.

# Последовательно преобразовать выборку в ранжированный, вариационный и интервальный ряды.
# Результаты содержательно проинтерпретировать.

# Ранжированный ряд

ranked_series = sample.sort_values(by = 'average_price')
ranked_series.rename(columns = {'average_price':'x_i'}, inplace = True)
ranked_series = ranked_series.values.tolist()


# В данном случае использую print, так как не получилось вывести индексы от 0 до n - 1 (они сортируются вместе с
# колонкой average_price)
print("Ранжированный ряд: ")
print("i  x_i")
for i in range(len(ranked_series)):
    strr = "{} {}".format(i, ranked_series[i])
    print(strr)
```

```python
# Вариационный ряд

variation_series = sample.groupby(sample.columns.tolist(), as_index = False).size()
variation_series.rename(columns = {'average_price':'x_i', 'size':'n_i'}, inplace = True)

print("\n\nВариационный ряд: ")
variation_series
```

```python
# Интервальный ряд

n = 95

max = variation_series['x_i'][len(variation_series) - 1]
min = variation_series['x_i'][0]

k = math.ceil(1 + math.log2(n))
R = round(max - min, 2)
h = R / k

first = []
last = []

first.append(min)

i = 1
x_curr = 0

while x_curr != max:
    
    x_curr = round(min + i * h, 2)
    
    if i == len(variation_series) - 1:
        last.append(x_curr)
    else:
        first.append(x_curr)
        last.append(x_curr)
    
    i += 1 

intervals = list(zip(first, last))
intervals_n = np.zeros(8)
intervals_x_mean = []

for i in range(len(variation_series)):
    
    x = variation_series['x_i'][i]
    n = variation_series['n_i'][i]

    if intervals[0][0] < x <= intervals[0][1]:
        intervals_n[0] += n
    elif x == intervals[0][0]:
        intervals_n[0] += n        
    elif intervals[1][0] < x <= intervals[1][1]:
        intervals_n[1] += n
    elif x == intervals[1][0]:
        intervals_n[1] += n 
    elif intervals[2][0] < x <= intervals[2][1]:
        intervals_n[2] += n
    elif x == intervals[2][0]:
        intervals_n[2] += n 
    elif intervals[3][0] < x <= intervals[3][1]:
        intervals_n[3] += n
    elif x == intervals[3][0]:
        intervals_n[3] += n 
    elif intervals[4][0] < x <= intervals[4][1]:
        intervals_n[4] += n
    elif x == intervals[4][0]:
        intervals_n[4] += n 
    elif intervals[5][0] < x <= intervals[5][1]:
        intervals_n[5] += n
    elif x == intervals[5][0]:
        intervals_n[5] += n 
    elif intervals[6][0] < x <= intervals[6][1]:
        intervals_n[6] += n
    elif x == intervals[6][0]:
        intervals_n[6] += n 
    elif intervals[7][0] < x <= intervals[7][1]:
        intervals_n[7] += n
    elif x == intervals[7][0]:
        intervals_n[7] += n 
        
for i in range(len(intervals)):
    x = round((intervals[i][0] + intervals[i][1]) / 2, 2)
    intervals_x_mean.append(x)
    
str_intervals = []
for i in range(len(intervals)):
    strr = "[{}, {})".format(intervals[i][0], intervals[i][1])
    str_intervals.append(strr)
    
table = pd.DataFrame(list(zip(str_intervals, intervals_x_mean, intervals_n)), columns = ['(x_i, x_i+1]', 
                                                                                         'x_ср',
                                                                                         'n_i'])

table
```

```python
# 3.

# Для интервального ряда абсолютных частот построить и отобразить графически полигон, гистограмму 
# и эмпирическую функцию. Сделать выводы.

sns.set_style("darkgrid")

# График полигона частот

print("Полигон абсолютных частот")
polygon = pd.DataFrame(list(zip(intervals_x_mean, intervals_n)), columns = ["x_m", "n_i"])
sns.lineplot(x = "x_m", y = "n_i", data = polygon)
```

```python
# График гистограммы

print("Гистограмма")
plt.hist(intervals_x_mean, weights = intervals_n, bins = k)
```

```python
# График эмпирической функции

comulative_n = np.zeros(8)
comulative_n[0] = intervals_n[0]

for i in range(1, len(intervals_n)):
    comulative_n[i] = comulative_n[i - 1] + intervals_n[i]

ls = sample.values.tolist()

bin_dt, bin_gr = np.histogram(ls, bins = k)

for i in range(len(comulative_n)):
    plt.plot([bin_gr[i], bin_gr[i + 1]], [comulative_n[i], comulative_n[i]], color = 'blue')  
    
print("\nБлагодаря полигону частот можно получить представление о распределении частот в выборке. Видно, что элемент, обладающий наибольшей частотой, находится в диапазоне от 1.1 до 1.2.")
print("\nПолигон и гистограмма относительных частот статистического распределения выборки аналогичны кривой распределения непрерывной случайной величины X, заданной плотностью вероятности f(x) = F'(x).")
```

```python
# 4.

# Аналогичные действия выполнить для интервального ряда относительных частот. 
# Сравнить результаты и сделать выводы.

# График полигона относительных частот

intervals_relative_n = []
n = 95

for i in range(len(intervals_n)):
    intervals_relative_n.append(intervals_n[i] / n)

print("Полигон абсолютных частот")
polygon = pd.DataFrame(list(zip(intervals_x_mean, intervals_relative_n)), columns = ["x_m", "p_i"])
sns.lineplot(x = "x_m", y = "p_i", data = polygon)
```

```python
# 5. Для интервального ряда найти середины интервалов, а также накопленные частоты. 
# Результаты представить в виде таблицы (в последней строке Σ необходимо записать сумму столбца).

sum_n = 0
sum_p = 0

for i in range(len(intervals_n)):
    sum_n += intervals_n[i]
    sum_p += intervals_relative_n[i]
    
table = pd.DataFrame(list(zip(str_intervals, 
                              intervals_x_mean, 
                              intervals_n,
                              intervals_relative_n, 
                              comulative_n,
                              comulative_rel_n,)), columns = ['(x_i, x_i+1]', 
                                                             'x_ср',
                                                             'n_i',
                                                             'p*_i',
                                                             'sum_n_i',
                                                             'sum_p*_i'])
 
table.loc['', :] = ['Сумма:', '-', sum_n, sum_p, '-', '-']
table
```

```python
# 6. Для полученных вариант вычислить условные варианты. Заполнить расчётную таблицу
# (в последней строке Σ необходимо записать сумму столбца). Проконтролировать корректность вычислений.

x_mean = intervals_x_mean

# В качестве ложного нуля выбираем x_mean[4], т.к. у x_mean[3] частота меньше
C = x_mean[4]

u = []

for i in range(len(x_mean)):
    u.append(round((x_mean[i] - C) / h, 3))

nu = []
nu_2 = []
nu_3 = []
nu_4 = []
n_u1_4 = []

for i in range(len(intervals_n)):    
    nu.append(intervals_n[i] * u[i])
    nu_2.append(intervals_n[i] * u[i]**2)
    nu_3.append(intervals_n[i] * u[i]**3)
    nu_4.append(intervals_n[i] * u[i]**4)
    
    u_ = u[i] + 1
    
    n_u1_4.append(intervals_n[i] * u_**4)

sum_n = intervals_n[0] + intervals_n[1] + intervals_n[2] + intervals_n[3] + intervals_n[4] + intervals_n[5] + intervals_n[6] + intervals_n[7]
sum_nu = nu[0] + nu[1] + nu[2] + nu[3] + nu[4] + nu[5] + nu[6] + nu[7]
sum_nu_2 = nu_2[0] + nu_2[1] + nu_2[2] + nu_2[3] + nu_2[4] + nu_2[5] + nu_2[6] + nu_2[7]
sum_nu_3 = nu_3[0] + nu_3[1] + nu_3[2] + nu_3[3] + nu_3[4] + nu_3[5] + nu_3[6] + nu_3[7]
sum_nu_4 = nu_4[0] + nu_4[1] + nu_4[2] + nu_4[3] + nu_4[4] + nu_4[5] + nu_4[6] + nu_4[7]
sum_n_u1_4 = n_u1_4[0] + n_u1_4[1] + n_u1_4[2] + n_u1_4[3] + n_u1_4[4] + n_u1_4[5] + n_u1_4[6] + n_u1_4[7]

control_sum = sum_n + 4 * sum_nu + 6 * sum_nu_2 + 4 * sum_nu_3 + sum_nu_4

table = pd.DataFrame(list(zip(intervals_x_mean, 
                              intervals_n, 
                              u,
                              nu, 
                              nu_2,
                              nu_3,
                              nu_4,
                              n_u1_4)), columns = ['x_i', 
                                                   'n_i',
                                                   'u_i',
                                                   'n_i*u_i',
                                                   'n_i*u_i^2',
                                                   'n_i*u_i^3',
                                                   'n_i*u_i^4',
                                                   'n_i(u_i + 1)^'])
 
table.loc['', :] = ['Сумма:', sum_n, '-', sum_nu, sum_nu_2, sum_nu_3, sum_nu_4, sum_n_u1_4]

strr = "\nКонтрольная сумма равна: {}. Она совпадает с суммой всех n_i(u_i + 1)^4. Следовательно, проверка пройдена.".format(control_sum)
print(strr)

table
```

```python
# 7. Вычислить условные эмпирические моменты 𝜈*_𝑖 через условные варианты. 
# С помощью условных эмпирических моментов вычислить центральные эмпирические моменты 𝜇*_𝑖 .
# Полученные результаты занести в таблицу.

n = 95

v_1 = (1 / n) * (intervals_n[0] * u[0] + intervals_n[1] * u[1] + intervals_n[2] * u[2] + intervals_n[3] * u[3] + intervals_n[4] * u[4] + intervals_n[5] * u[5] + intervals_n[6] * u[6] + intervals_n[7] * u[7])
v_2 = (1 / n) * (intervals_n[0] * u[0]**2 + intervals_n[1] * u[1]**2 + intervals_n[2] * u[2]**2 + intervals_n[3] * u[3]**2 + intervals_n[4] * u[4]**2 + intervals_n[5] * u[5]**2 + intervals_n[6] * u[6]**2 + intervals_n[7] * u[7]**2)
v_3 = (1 / n) * (intervals_n[0] * u[0]**3 + intervals_n[1] * u[1]**3 + intervals_n[2] * u[2]**3 + intervals_n[3] * u[3]**3 + intervals_n[4] * u[4]**3 + intervals_n[5] * u[5]**3 + intervals_n[6] * u[6]**3 + intervals_n[7] * u[7]**3)
v_4 = (1 / n) * (intervals_n[0] * u[0]**4 + intervals_n[1] * u[1]**4 + intervals_n[2] * u[2]**4 + intervals_n[3] * u[3]**4 + intervals_n[4] * u[4]**4 + intervals_n[5] * u[5]**4 + intervals_n[6] * u[6]**4 + intervals_n[7] * u[7]**4)

mu_1 = 0
mu_2 = (v_2 - v_1**2) * h**2
mu_3 = (v_3 - 3 * v_1 * v_2 + 2 * v_1**3) * h**3
mu_4 = (v_4 - 4 * v_1 * v_3 + 6 * v_2 * v_1**2 - 3 * v_1**4) * h**4

table = pd.DataFrame(list(zip([v_1, v_2, v_3, v_4], 
                              [mu_1, mu_2, mu_3, mu_4])), columns = ['ν*_i', 'μ*_i'])
 
table
```
