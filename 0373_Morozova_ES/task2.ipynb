#### –§–æ—Ä–º—É–ª—ã –ø–µ—Ä–≤—ã—Ö —á–µ—Ç—ã—Ä—ë—Ö —Ü–µ–Ω—Ç—Ä–∞–ª—å–Ω—ã—Ö —ç–º–ø–∏—Ä–∏—á–µ—Å–∫–∏—Ö –º–æ–º–µ–Ω—Ç–æ–≤:

#### –§–æ—Ä–º—É–ª—ã –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ –∞—Å–∏–º–º–µ—Ç—Ä–∏–∏ –∏ —ç–∫—Å—Ü–µ—Å—Å–∞:

#### –§–æ—Ä–º—É–ª–∞ –º–æ–¥—ã:

#### –§–æ—Ä–º—É–ª–∞ –º–µ–¥–∏–∞–Ω—ã:

#### –§–æ—Ä–º—É–ª–∞ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞ –≤–∞—Ä–∏–∞—Ü–∏–∏:

### –î–ª—è –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω–æ–≥–æ –∏ —Ä–µ–≥—Ä–µ—Å—Å–∏–æ–Ω–Ω–æ–≥–æ –∞–Ω–∞–ª–∏–∑–∞:
  
#### –§–æ—Ä–º—É–ª–∞ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏:

$ r_B = \frac{\frac{1}{n}\sum_{i = 1}^{k_u}\sum_{j = 1}^{k_v}n_{uv}u_iv_j - \bar{u}\bar{v}}{\sigma_u\sigma_v} $  
  
$ \sum_{i = 1}^{k_u}\sum_{j = 1}^{k_v}n_{uv}u_iv_j = \sum_{j = 1}^{k_v}v_jU_j, U_j = \sum_{i = 1}^{k_u}n_{uv}u_i $  
$ \sum_{i = 1}^{k_u}\sum_{j = 1}^{k_v}n_{uv}u_iv_j = \sum_{i = 1}^{k_u}u_iV_i, V_i = \sum_{j = 1}^{k_v}n_{uv}v_j $  
#### –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞ –¥–ª—è –æ—Ü–µ–Ω–∫–∏ –≤—ã–±–æ—Ä–æ—á–Ω–æ–≥–æ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏:

$ (z - \frac{t}{\sqrt{n - 3}}; z + \frac{t}{\sqrt{n - 3}}), \Phi(t) = \frac{\gamma}{2} $ - –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª –¥–ª—è —Å.–≤. z.  

$ (th(z_l); th(z_u)) $ - –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª –¥–ª—è –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏.  
  
#### –ü—Ä–æ–≤–µ—Ä–∫–∞ –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏:

–û–±–ª–∞—Å—Ç—å –ø—Ä–∏–Ω—è—Ç–∏—è –≥–∏–ø–æ—Ç–µ–∑—ã: $ \left| T_{nabl} \right| < T_{krit} $  

#### –£—Ä–∞–≤–Ω–µ–Ω–∏–µ —Ä–µ–≥—Ä–µ—Å—Å–∏–∏:  
$ y - \bar{y_B} = r_B\frac{\sigma_y}{\sigma_x}(x - \bar{x_B}) $  
$ x - \bar{x_B} = r_B\frac{\sigma_x}{\sigma_y}(y - \bar{y_B}) $  
  
#### –§–æ—Ä–º—É–ª—è –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω–æ–≥–æ –æ—Ç–Ω–æ—à–µ–Ω–∏—è:  
$ \eta_{yx} = \frac{\sigma_{mezhgr}}{\sigma_{ob}} $  
$ \eta_{yx} = \frac{\sigma_{\bar{y}x}}{\sigma_y} $  
  
$ \sigma_{\bar{y}x} = \sqrt{\sigma^2_{mezhgr}} = \sqrt{\sum n_x(\bar{y_x} - \bar{y})^2 / n} $  
$ \sigma_{y} = \sqrt{\sigma^2_{ob}} = \sqrt{\sum n_y(y - \bar{y})^2 / n} $  
  
$ \eta_{xy} = \frac{\sigma_{mezhgr}}{\sigma_{ob}} $  
$ \eta_{xy} = \frac{\sigma_{\bar{x}y}}{\sigma_x} $  
  
$ \sigma_{\bar{x}y} = \sqrt{\sigma^2_{mezhgr}} = \sqrt{\sum n_y(\bar{x_y} - \bar{x})^2 / n} $  
$ \sigma_{x} = \sqrt{\sigma^2_{ob}} = \sqrt{\sum n_x(x - \bar{x})^2 / n} $

#### –ü–æ–∏—Å–∫ —É—Ä–∞–≤–Ω–µ–Ω–∏—è –ø–∞—Ä–∞–±–æ–ª–∏—á–µ—Å–∫–æ–π –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏  

$ y = \beta_0 + \beta_1x + \beta_2x^2 $  
  
$ \beta_0, \beta_1, \beta_2 $ –Ω–∞—Ö–æ–¥—è—Ç—Å—è –ø–æ –º–µ—Ç–æ–¥—É –Ω–∞–∏–º–µ–Ω—å—à–∏—Ö –∫–≤–∞–¥—Ä–∞—Ç–æ–≤.  
  
$ \left\{\begin{matrix} \frac{dF}{d\beta_0} = -2\sum_{i = 1}^{n}(y_i - \beta_0 - \beta_1x_i -\beta_2x^2_i) = 0;
 \\ \frac{dF}{d\beta_1} = -2\sum_{i = 1}^{n}(y_i - \beta_0 - \beta_1x_i -\beta_2x^2_i)x_i = 0;
 \\ \frac{dF}{d\beta_2} = -2\sum_{i = 1}^{n}(y_i - \beta_0 - \beta_1x_i -\beta_2x^2_i)x^2_i = 0;
\end{matrix}\right. $  
  
$ \left\{\begin{matrix} \beta_0n + \beta_1\sum_{i = 1}^{n}x_i + \beta_2\sum_{i = 1}^{n}x^2_i = \sum_{i = 1}^{n}y_i,
 \\ \beta_0\sum_{i = 1}^{n}x_i + \beta_1\sum_{i = 1}^{n}x^2_i + \beta_2\sum_{i = 1}^{n}x^3_i = \sum_{i = 1}^{n}(x_iy_i),
 \\ \beta_0\sum_{i = 1}^{n}x^2_i + \beta_1\sum_{i = 1}^{n}x^3_i + \beta_2\sum_{i = 1}^{n}x^4_i = \sum_{i = 1}^{n}(x^2_iy_i).
\end{matrix}\right. $  
  
$ S_1 = \sum_{i = 1}^{n}x_i $  
$ S_2 = \sum_{i = 1}^{n}x^2_i $  
$ S_3 = \sum_{i = 1}^{n}x^3_i $  
$ S_4 = \sum_{i = 1}^{n}x^4_i $  
$ S_5 = \sum_{i = 1}^{n}y_i $  
$ S_6 = \sum_{i = 1}^{n}x_iy_i $  
$ S_7 = \sum_{i = 1}^{n}x^2_iy_i $  
  
–ù–µ–∏–∑–≤–µ—Å—Ç–Ω—ã–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã —Å–∏—Å—Ç–µ–º—ã –Ω–∞—Ö–æ–¥—è—Ç—Å—è –º–µ—Ç–æ–¥–æ–º –ö—Ä–∞–º–µ—Ä–∞:  
$ \Delta = \begin{bmatrix}

\end{bmatrix}  $

#### –ü–æ–∏—Å–∫ —É—Ä–∞–≤–Ω–µ–Ω–∏—è —ç–∫—Å–ø–æ–Ω–µ–Ω—Ü–∏–∞–ª—å–Ω–æ–π –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏  
$ y = \beta_0 e^{\beta_1x} $  
–ü–æ—Å–ª–µ –ª–∏–Ω–µ–∞—Ä–∏–∑–∞—Ü–∏–∏: $ ln(y) = ln(B_0) + B_1x $  
  
–ù–∞—Ö–æ–¥–∏–º –Ω–µ–∏–∑–≤–µ—Å—Ç–Ω—ã–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã —Å –ø–æ–º–æ—â—å—é –ú–ù–ö:  
$ \left\{\begin{matrix}B_0n + B_1\sum_{i = 1}^{n}x_i = \sum_{i = 1}^{n}y_i

$ S_1 = \sum_{i = 1}^{n}x_i $

$ \left\{\begin{matrix}B_0 = \frac{S_3 - S_1B_1}{n}

–ù–∞—Ö–æ–¥–∏–º $ B_1 $, –ø–æ–¥—Å—Ç–∞–≤–ª—è–µ–º –≤ $ B_0 $ –∏ –Ω–∞—Ö–æ–¥–∏–º –∏—Å–∫–æ–º—ã–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –∫–∞–∫ $ \beta_0 = B_0, \beta_1 = e^{B_1} $.  
  
#### –§–æ—Ä–º—É–ª—ã –ø–æ–∫–∞–∑–∞—Ç–µ–ª–µ–π –∫–∞—á–µ—Å—Ç–≤–∞ (–∞–¥–µ–∫–≤–∞—Ç–Ω–æ—Å—Ç–∏) —Ä–µ–≥—Ä–µ—Å—Å–∏–∏  
–û—Å—Ç–∞—Ç–∫–∏: $ \varepsilon_{y_i} = y_i - y_{x_i}, \varepsilon_{x_j} = x_j - x_{y_j} $  
  
–†–∞–∑–ª–æ–∂–µ–Ω–∏–µ –¥–∏—Å–ø–µ—Ä—Å–∏–∏: $ \sigma^2 = (\delta^*)^2 + (\varepsilon^*)^2 $, –≥–¥–µ  
$ (\delta^*_y)^2 = \frac{1}{n}\sum_{i = 1}^{n}(y_{x_i} - \bar{y})^2, (\delta^*_x)^2 = \frac{1}{n}\sum_{i = 1}^{n}(x_{y_i} - \bar{x})^2 $ - –æ–±—ä—è—Å–Ω–µ–Ω–Ω–∞—è —É—Ä–∞–≤–Ω–µ–Ω–∏–µ–º —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ –¥–∏—Å–ø–µ—Ä—Å–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∏–≤–Ω–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞;

–¢–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏–π –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –¥–µ—Ç–µ—Ä–º–∏–Ω–∞—Ü–∏–∏:

$ S_{\varepsilon_y} = \sqrt{\frac{1}{n - l}\sum_{i = 1}^{n}\varepsilon^2_{y_i}}, S_{\varepsilon_x} = \sqrt{\frac{1}{n - l}\sum_{i = 1}^{n}\varepsilon^2_{x_i}} $, –≥–¥–µ l - —á–∏—Å–ª–æ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –≤ –º–æ–¥–µ–ª–∏ —Ä–µ–≥—Ä–µ—Å—Å–∏–∏

$ A_y = \frac{1}{n}\sum_{i = 1}^{n}\left| \frac{\varepsilon_{y_i}}{y_i} \right|, A_x = \frac{1}{n}\sum_{j = 1}^{n}\left| \frac{\varepsilon_{x_j}}{x_j} \right| $
