## Основные теоретические положения

### Для статистической обработки второй выборки:

#### Формулы для построения интервального ряда:  

По формуле Старджесса находим оптимальное количество интервалов: $ k = 1 + \log_2 n $  
Длина каждого интервала равна: $ h = \frac{y_{max} - y_{min}}{k} $  
$ y_{0} = y_{min} $  
$ y_{i} = y_{0} + ih $  

#### Формула середины интервалов:
$ \tilde{y} = \frac{y_{i - 1} + y_{i}}{2} $  

#### Формула относительной частоты:
$ p^{*}_{i} = \frac{n_{i}}{n} $  
  
#### Формула выборочного среднего:  
$ \bar{y}_{B} = \frac{1}{n}\sum^n_{i + 1}y_in_i $  
  
#### Формула выборочной дисперсии:  
$ \sigma^2_{B} = \frac{1}{n}\sum^n_{i = 1}(y_i - \bar{y}_B)^2 $  

#### Формула исправленных выборочных дисперсии и СКО:
$ s^2 = \frac{n}{n - 1}\sigma^2_B = \frac{1}{n - 1}\sum^k_{i = 1}n_i(x_i - \bar{y_B})^2 $  
$ s = \sqrt{s^2} $  

#### Формула условной варианты:
  
$ u_{i} = \frac{y_{i} - C}{h} $, где $ C $ - ложный нуль, $ h $ - шаг  
Ложный нуль выбирается как варианта, находящаяся посередине. Если количество интервалов чётное, то выбирается варианта с наибольшей частотой.  
  
#### Формула условного эмпирического момента:  
$ \nu^*_{r} = \frac{1}{n}\sum^k_{i = 1}n_iu^r_i $  

#### Формулы первых четырёх центральных эмпирических моментов:  
$ \mu^*_{3} = (\nu^*_{3} - 3\nu^*_{1}\nu^*_{2} + 2(\nu^*_{1})^3)h^3 $  
$ \mu^*_{4} = (\nu^*_{4} - 4\nu^*_{1}\nu^*_{3} + 6\nu^*_{2}(\nu^*_{1})^2 - 3(\nu^*_{1})^4)h^4 $  
  
#### Формулы коэффициентов асимметрии и эксцесса:
$ a^*_s = \frac{\mu^*_3}{\sigma^3_B} $  
$ \varepsilon^*_k = \frac{\mu^*_4}{\sigma^4_B} - 3 $  
  
#### Формула моды:  
$ M^*_o \in (y^o_{M_o}, y^{'}_{M_o}); M^*_o = y^o_{M_o} + h\frac{n_{M_o} - n_{M_o - 1}}{(n_{M_o} - n_{M_o - 1}) + (n_{M_o} - n_{M_o + 1})} $  

#### Формула медианы:   
$ M^*_e \in (y^o_{M_e}, y^{'}_{M_e}); M^*_e = y^o_{M_e} + \frac{h}{p^*_{M_e}}(0.5 - p^{\sum}_{M_e - 1}) $  
  
#### Формула коэффициента вариации:   
$ \upsilon^* = \frac{\sigma_B}{|\bar{Y_B}|}*100  $  

### Для корреляционного и регрессионного анализа:
  
#### Формула коэффициента корреляции:  
$ u_i = \frac{x_i - C_x}{h_x} $  
$ v_i = \frac{y_i - C_y}{h_y} $  
  
$ r_B = \frac{\frac{1}{n}\sum_{i = 1}^{k_u}\sum_{j = 1}^{k_v}n_{uv}u_iv_j - \bar{u}\bar{v}}{\sigma_u\sigma_v} $  
  
$ \sum_{i = 1}^{k_u}\sum_{j = 1}^{k_v}n_{uv}u_iv_j = \sum_{j = 1}^{k_v}v_jU_j, U_j = \sum_{i = 1}^{k_u}n_{uv}u_i $  
$ \sum_{i = 1}^{k_u}\sum_{j = 1}^{k_v}n_{uv}u_iv_j = \sum_{i = 1}^{k_u}u_iV_i, V_i = \sum_{j = 1}^{k_v}n_{uv}v_j $  
#### Построение доверительного интервала для оценки выборочного коэффициента корреляции:
$ z = arth(r_B) $ - преобразование Фишера. 

$ (z - \frac{t}{\sqrt{n - 3}}; z + \frac{t}{\sqrt{n - 3}}), \Phi(t) = \frac{\gamma}{2} $ - доверительный интервал для с.в. z.  

$ (th(z_l); th(z_u)) $ - доверительный интервал для коэффициента корреляции.  
  
#### Проверка значимости коэффициента корреляции:  
$ H_0: r_B = 0 $  
$ H_1: r_B \neq 0 $  
  
$ T_{nabl} = r_b\frac{\sqrt{n - 2}}{\sqrt{1 - r^{2}_B}} $  
$ T_{krit} = t(\alpha, df), df = n - 2 $  
  
Область принятия гипотезы: $ \left| T_{nabl} \right| < T_{krit} $  

#### Уравнение регрессии:  
$ y - \bar{y_B} = r_B\frac{\sigma_y}{\sigma_x}(x - \bar{x_B}) $  
$ x - \bar{x_B} = r_B\frac{\sigma_x}{\sigma_y}(y - \bar{y_B}) $  
  
#### Формуля корреляционного отношения:  
$ \eta_{yx} = \frac{\sigma_{mezhgr}}{\sigma_{ob}} $  
$ \eta_{yx} = \frac{\sigma_{\bar{y}x}}{\sigma_y} $  
  
$ \sigma_{\bar{y}x} = \sqrt{\sigma^2_{mezhgr}} = \sqrt{\sum n_x(\bar{y_x} - \bar{y})^2 / n} $  
$ \sigma_{y} = \sqrt{\sigma^2_{ob}} = \sqrt{\sum n_y(y - \bar{y})^2 / n} $  
  
$ \eta_{xy} = \frac{\sigma_{mezhgr}}{\sigma_{ob}} $  
$ \eta_{xy} = \frac{\sigma_{\bar{x}y}}{\sigma_x} $  
  
$ \sigma_{\bar{x}y} = \sqrt{\sigma^2_{mezhgr}} = \sqrt{\sum n_y(\bar{x_y} - \bar{x})^2 / n} $  
$ \sigma_{x} = \sqrt{\sigma^2_{ob}} = \sqrt{\sum n_x(x - \bar{x})^2 / n} $

#### Поиск уравнения параболической зависимости  

$ y = \beta_0 + \beta_1x + \beta_2x^2 $  
  
$ \beta_0, \beta_1, \beta_2 $ находятся по методу наименьших квадратов.  
  
$ \left\{\begin{matrix} \frac{dF}{d\beta_0} = -2\sum_{i = 1}^{n}(y_i - \beta_0 - \beta_1x_i -\beta_2x^2_i) = 0;
 \\ \frac{dF}{d\beta_1} = -2\sum_{i = 1}^{n}(y_i - \beta_0 - \beta_1x_i -\beta_2x^2_i)x_i = 0;
 \\ \frac{dF}{d\beta_2} = -2\sum_{i = 1}^{n}(y_i - \beta_0 - \beta_1x_i -\beta_2x^2_i)x^2_i = 0;
\end{matrix}\right. $  
  
$ \left\{\begin{matrix} \beta_0n + \beta_1\sum_{i = 1}^{n}x_i + \beta_2\sum_{i = 1}^{n}x^2_i = \sum_{i = 1}^{n}y_i,
 \\ \beta_0\sum_{i = 1}^{n}x_i + \beta_1\sum_{i = 1}^{n}x^2_i + \beta_2\sum_{i = 1}^{n}x^3_i = \sum_{i = 1}^{n}(x_iy_i),
 \\ \beta_0\sum_{i = 1}^{n}x^2_i + \beta_1\sum_{i = 1}^{n}x^3_i + \beta_2\sum_{i = 1}^{n}x^4_i = \sum_{i = 1}^{n}(x^2_iy_i).
\end{matrix}\right. $  
  
$ S_1 = \sum_{i = 1}^{n}x_i $  
$ S_2 = \sum_{i = 1}^{n}x^2_i $  
$ S_3 = \sum_{i = 1}^{n}x^3_i $  
$ S_4 = \sum_{i = 1}^{n}x^4_i $  
$ S_5 = \sum_{i = 1}^{n}y_i $  
$ S_6 = \sum_{i = 1}^{n}x_iy_i $  
$ S_7 = \sum_{i = 1}^{n}x^2_iy_i $  
  
Неизвестные коэффициенты системы находятся методом Крамера:  
$ \Delta = \begin{bmatrix}
n & S_1 & S_2 \\
S_1 & S_2 & S_3 \\
S_2 & S_3 & S_4 \\
\end{bmatrix}  $  
  
$ \Delta_0 = \begin{bmatrix}
S_5 & S_1 & S_2 \\
S_6 & S_2 & S_3 \\
S_7 & S_3 & S_4 \\
\end{bmatrix}  $  
  
$ \Delta_1 = \begin{bmatrix}
n & S_5 & S_2 \\
S_1 & S_6 & S_3 \\
S_2 & S_7 & S_4 \\
\end{bmatrix}  $  
  
$ \Delta_2 = \begin{bmatrix}
n & S_1 & S_5 \\
S_1 & S_2 & S_6 \\
S_2 & S_3 & S_7 \\
\end{bmatrix}  $  
  
$ \beta_0 = \frac{\Delta_0}{\Delta}, \beta_1 = \frac{\Delta_1}{\Delta}, \beta_2 = \frac{\Delta_2}{\Delta}. $  
  
#### Поиск уравнения экспоненциальной зависимости  
$ y = \beta_0 e^{\beta_1x} $  
После линеаризации: $ ln(y) = ln(B_0) + B_1x $  
  
Находим неизвестные параметры с помощью МНК:  
$ \left\{\begin{matrix}B_0n + B_1\sum_{i = 1}^{n}x_i = \sum_{i = 1}^{n}y_i
 \\ B_0\sum_{i = 1}^{n}x_i + B_1\sum_{i = 1}^{n}x^2_i = \sum_{i = 1}^{n}y_ix_i
\end{matrix}\right. $  
  
$ S_1 = \sum_{i = 1}^{n}x_i $  
$ S_2 = \sum_{i = 1}^{n}x^2_i $  
$ S_3 = \sum_{i = 1}^{n}y_i $  
$ S_4 = \sum_{i = 1}^{n}y_ix_i $  
  
$ \left\{\begin{matrix}B_0 = \frac{S_3 - S_1B_1}{n}
 \\ B_1 = \frac{nS_4 - S_1S_3}{S_2n - S^2_1}
\end{matrix}\right. $  
Находим $ B_1 $, подставляем в $ B_0 $ и находим искомые коэффициенты как $ \beta_0 = B_0, \beta_1 = e^{B_1} $.  
  
#### Формулы показателей качества (адекватности) регрессии  
Остатки: $ \varepsilon_{y_i} = y_i - y_{x_i}, \varepsilon_{x_j} = x_j - x_{y_j} $  
  
Разложение дисперсии: $ \sigma^2 = (\delta^*)^2 + (\varepsilon^*)^2 $, где  
$ (\delta^*_y)^2 = \frac{1}{n}\sum_{i = 1}^{n}(y_{x_i} - \bar{y})^2, (\delta^*_x)^2 = \frac{1}{n}\sum_{i = 1}^{n}(x_{y_i} - \bar{x})^2 $ - объясненная уравнением регрессии дисперсия результативного признака;  
$ (\varepsilon^*_y)^2 = \frac{1}{n}\sum_{i = 1}^{n}\varepsilon^2_y, (\varepsilon^*_x)^2 = \frac{1}{n}\sum_{i = 1}^{n}\varepsilon^2_x $ - необъясненная уравнением регрессии (остаточная) дисперсия.  
  
Теоретический коэффициент детерминации:  
$ R^2_y = \frac{(\delta^*_y)^2}{\sigma^2_y}, R^2_x = \frac{(\delta^*_x)^2}{\sigma^2_x} $  
  
Средняя квадратическая ошибка:  
$ S_{\varepsilon_y} = \sqrt{\frac{1}{n - l}\sum_{i = 1}^{n}\varepsilon^2_{y_i}}, S_{\varepsilon_x} = \sqrt{\frac{1}{n - l}\sum_{i = 1}^{n}\varepsilon^2_{x_i}} $, где l - число параметров в модели регрессии  
  
Средняя ошибка аппроксимации:  
$ A_y = \frac{1}{n}\sum_{i = 1}^{n}\left| \frac{\varepsilon_{y_i}}{y_i} \right|, A_x = \frac{1}{n}\sum_{j = 1}^{n}\left| \frac{\varepsilon_{x_j}}{x_j} \right| $
