## Основные теоретические положения
Кластерный анализ - задача разбиения заданной выборки объектов на подмножества, называемые кластерами так, чтобы каждый кластер состоял из схожих объектов, а объекты разных кластеров существенно отличались.

Задача кластеризации относится к статистической обработке, а также широкому классу задач, обучающих без учителя.

Кластерный анализ - классификация объектов в относительно однородные группы(кластеры).

Основные задачи:
 разработка типологииили классификации,
 исследование полезных схем группировки объектов,
 порождение гипотез на основе исследований,
 проверка гипотез для определения, действительно ли группы, выделенные каким-либо методом, есть в имеющихся данных.

Основные цели:
 получение информации о данных, их понимание при помощи выявления кластерной структуры,
 сжатие данных,
 выделение нетипичных объектов.

Этапы кластеризации:
 отбор выборки,
 определение множества переменных, по которому будут оцениваться объекты выборки. Нормализация переменных,
 вычисление значений меры сходства между объектами,
 применение метода кластерного анализа для создания групп сходных кластеров,
 представление анализа.

Метрика (функция расстояния) - неотрицательная вещественная функция, определена на некотором метрическом пространстве, если выполняется аксиома неотрицательности d(x,y)>=0, аксиома коммутативности d(x,y)=d(y,x), правило треугольника d(x,y)<=d(x,z)+d(z,y).

Функции расстояния (метрики):

Евклидова метрика: $d(x,y) = \sqrt{\sum_{i=1}^{n}(x_i-y_i)^2}$

Квадрат Евклидовой метрики: $d(x,y) = \sum_{i=1}^{n}(x_i-y_i)^2$

Обобщенное степенное расстояние Минковского:  $d(x,y) = (\sum_{i=1}^{n}(x_i-y_i)^p)^{\frac{1}{p}}$

Расстояние Чебышева: $d(x,y) = \max_{i=[1,n]}(x_i-y_i)$

Манхэттенское расстояние:  $d(x,y) = \sum_{i=1}^{n}|(x_i-y_i)|$

Метод k-means: суть - минимизация суммарного квадратичного отклонения точек кластеров от центров этих кластеров.
Алгоритм:Случайно выбираются k точек, являющиеся начальными центрами кластеров

Относим каждый объект к кластеру с ближайщим центроидом

Перестраиваем центроиды

Если критерий остановки неудовлетворен, то возвращаемся к пункту 2

Метод k-medians - вариация k-means, где для определения центра кластера вычисляется медиана по каждому измерению. Предназанчени для более качественной кластеризации, если  k-means не справился. Для рассчета расстояния берем Манхэттенское расстояние.

Поиск грубой оценки верхнего колличества кластеров: $k = \sqrt{\frac{n}{2}}$,  округляем в меньшую сторону.

Метод силуэтов - Коэффициент силуэта вычисляется с помощью внутрикластерного расстояния и среднего расстояния до ближайшего кластера по каждому кластеру. $S_i = \frac{b_i-a_i}{\max(a_i;b_i)}$, где $a_i = \frac{1}{\#C_j-1}\sum_{x_j \in C_i}d(x_j,x_{C_j})$; $b_i = \min_{j \ne i}\frac{1}{\#C_j}\sum_{x_j \in C_j}d(x_j,x_{C_i})$


Метод поиска сгущений: Заключается в разбиении множества на кластеры, не знаем их количество, но знаем их радиус.
Алгоритм: Инициализируем множество необработанных точек

выбираем случайную точку C, так называемый центроид

раздуваем вокруг нее сферу с радиусом R и смотрим, какие точки в нее попали. Образовали кластер.

перемещаем центроид: $\overline{x_C} = \frac{1}{\# K}\sum_{x \in K}x$

делаем проверку. Если предыдущий и новый центроиды не совпали, то перемещяем центр и переходим к пункту 3

исключаем из множества точки, которые уже в кластерах

добавляем к кластеру центроид

Если еще есть необработанные точки, то переходим к пункту 2, иначе заканчиваем.

Проверка чувствительности метода: берем радиус, считаем кластеризацию с небольшой погрешностью $\overline{R}=R+\epsilon$ и смотрим, как изменится кластеризация.

### Задание 3
Реализовать алгоритм k-means. Отобразить полученные кластеры,
каждый кластер выделить разным цветом или маркером, отметить
центроиды

```python
k = 7 #нашли в п.2
XY = np.column_stack((X1,Y1))
n=101
COLORS = ('green', 'blue', 'red', 'black', 'yellow', 'brown', 'pink', 'orange', 'purple', 'gray')
def evclid(x, y): # квадрат евклидовой метрики
    h = np.sqrt((x[0] - y[0])**2 + (x[1] - y[1])**2)
    return h
def k_means (n, k):
    cluster = {}
    cluster = [[] for i in range (7)]
    size, c = [], []
    centr = {}
    index_centr = [10, 25, 45, 50, 75, 90, 100]
    for i in range (k):
        centr[i] = [XY[index_centr[i]][0], XY[index_centr[i]][1]] 
    pred_centr = copy.deepcopy(centr)
    change = True
    while (change):
        change = False
        for i in range(101):
            distance = []
            for j in range (k):
                distance.append(evclid(XY[i], centr[j]))
            cluster[np.argmin(distance)].append(XY[i])
        for i in range(k):
            if (len(cluster[i]) != 0): centr[i]=[np.mean(np.array(cluster[i])[:,0]),np.mean(np.array(cluster[i])[:,1])]
            if (pred_centr == centr): change = True
        pred_centr = copy.deepcopy(centr)
    for i in range (k):
        size.append(len(cluster[i]))
    for i in range (k):
        c.append([round(centr[i][0], 3), round(centr[i][1], 3)])
    return cluster, centr, size, c
def method(k, clusters, centr):
    plt.figure(figsize = (10, 10))
    for i in range(k):
        for j in range(len(clusters[i])):
            plt.scatter(clusters[i][j][0], clusters[i][j][1], color= COLORS[i], s=10)
        plt.scatter(round(centr[i][0], 3), round(centr[i][1], 3), color= COLORS[i],  s=50)
    plt.grid(True)   
    plt.show()   
    
for i in range(1, 8):
    cluster, centr, size_means, c_means = k_means(n,i)
    print(method(i, cluster, centr))
```

### Задание 4
Реализовать алгоритмы k-medians. Отобразить полученные кластеры, каждый кластер выделить разным цветом или маркером,
отметить центроиды.

```python
def manhattan (x, y):
    h2 = abs(x[0] - y[0]) + abs(x[1] - y[1])
    return h2
def kmedian (n, k):
    cluster = {}
    cluster = [[] for i in range (k)]
    centr = {}
    size, c = [], []
    index_centr = [10, 25, 45, 50, 75, 90, 100]
    for i in range (k):
        centr[i] = [XY[index_centr[i]][0], XY[index_centr[i]][1]] 
    pred_centr = copy.deepcopy(centr)
    change = True
    while (change):
        change = False
        for i in range(101):
            distance = []
            for j in range (k):
                distance.append(manhattan(XY[i], centr[j]))
            cluster[np.argmin(distance)].append(XY[i])
        for i in range(k):
            if (len(cluster[i]) != 0): centr[i]=[np.median(np.array(cluster[i])[:,0]),np.median(np.array(cluster[i])[:,1])]
            if (pred_centr == centr): change = True
        pred_centr = copy.deepcopy(centr)
    for i in range (k):
        size.append(len(cluster[i]))
    for i in range (k):
        c.append([np.round(centr[i][0], 3), np.round(centr[i][1], 3)])
    return cluster, centr, size, c   
for i in range(1, 8):
    cluster, centr, size_m, c_m = kmedian(n,i)
    print(method(i, cluster, centr))
```

### Задание 7
 Реализовать алгоритм поиска сгущений при различных радиусах
𝑅, отобразить полученные кластеры, выделить каждый кластер
разным цветом или маркером, отметить центроиды. Результаты
представить в виде таблицы:

```python
R_max = -10000
R_min = 10000
def FOREL(R):
    u = [[XY[i][0], XY[i][1]] for i in range (n)]
    cluster, centr = {}, {}
    k = 0
    while(len(u) > 0):
        change = True
        xc = u[0]
        while change:
            u0, u1 = 0, 0
            cluster[k]=[]
            for i in range(len(u)):
                if evclid(xc, u[i]) <= R:
                    u0 += u[i][0]
                    u1 += u[i][1]
                    cluster[k].append(u[i])
            new_xc = xc
            xc = [1/(len(cluster[k])) * u0, 1 / len(cluster[k]) * u1]
    
            if xc == new_xc:
                centr[k] = xc
                change = False             
        for i in cluster[k]:
            if i in u:
                u.remove(i)
        k = k + 1
    size = []
    c = []
    for i in range (k):
        size.append(len(cluster[i]))
    for i in range (k):
        c.append([round(centr[i][0], 4), round(centr[i][1], 4)])
    return k, cluster, centr, size, c
for i in range (n-1):
    for j in range (i+1, n):
            d = evclid(XY[i],XY[j])
            if d > 0:
                if R_max < d:
                    R_max = d
                if R_min > d:
                    R_min = d
print('R_max=', R_max)
print('R_min=', R_min)
def forel(k, clusters, centr):
    plt.figure(figsize = (10,10))
    for i in range(k):
        for j in range(len(clusters[i])):
            plt.scatter(clusters[i][j][0], clusters[i][j][1], color= COLORS[i], s=10)
        plt.scatter(round(centr[i][0], 4), round(centr[i][1], 4), s=50, color= COLORS[i])
    plt.grid(True)
    plt.show()
R = [1.5, 2, 2.5, 3, 5.5]
for i in range(len(R)):
    k, cluster, centr, size, c = FOREL(R[i])
    forel(k, cluster, centr) 
size, c = [], []
for i in range(len(R)):
    size.append(FOREL(R[i])[3])
for i in range(len(R)):
    c.append(FOREL(R[i])[4])    
count3 = pd.DataFrame({'R': R, "Центры": c, 'размерность': size})
print(count3)
```
