### –ó–∞–¥–∞–Ω–∏–µ 3
–†–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º k-means. –û—Ç–æ–±—Ä–∞–∑–∏—Ç—å –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ –∫–ª–∞—Å—Ç–µ—Ä—ã,
–∫–∞–∂–¥—ã–π –∫–ª–∞—Å—Ç–µ—Ä –≤—ã–¥–µ–ª–∏—Ç—å —Ä–∞–∑–Ω—ã–º —Ü–≤–µ—Ç–æ–º –∏–ª–∏ –º–∞—Ä–∫–µ—Ä–æ–º, –æ—Ç–º–µ—Ç–∏—Ç—å
—Ü–µ–Ω—Ç—Ä–æ–∏–¥—ã

```python
k = 7 #–Ω–∞—à–ª–∏ –≤ –ø.2
XY = np.column_stack((X1,Y1))
n=101
COLORS = ('green', 'blue', 'red', 'black', 'yellow', 'brown', 'pink', 'orange', 'purple', 'gray')
def evclid(x, y): # –∫–≤–∞–¥—Ä–∞—Ç –µ–≤–∫–ª–∏–¥–æ–≤–æ–π –º–µ—Ç—Ä–∏–∫–∏
    h = np.sqrt((x[0] - y[0])**2 + (x[1] - y[1])**2)
    return h
def k_means (n, k):
    cluster = {}
    cluster = [[] for i in range (7)]
    size, c = [], []
    centr = {}
    index_centr = [10, 25, 45, 50, 75, 90, 100]
    for i in range (k):
        centr[i] = [XY[index_centr[i]][0], XY[index_centr[i]][1]] 
    pred_centr = copy.deepcopy(centr)
    change = True
    while (change):
        change = False
        for i in range(101):
            distance = []
            for j in range (k):
                distance.append(evclid(XY[i], centr[j]))
            cluster[np.argmin(distance)].append(XY[i])
        for i in range(k):
            if (len(cluster[i]) != 0): centr[i]=[np.mean(np.array(cluster[i])[:,0]),np.mean(np.array(cluster[i])[:,1])]
            if (pred_centr == centr): change = True
        pred_centr = copy.deepcopy(centr)
    for i in range (k):
        size.append(len(cluster[i]))
    for i in range (k):
        c.append([round(centr[i][0], 3), round(centr[i][1], 3)])
    return cluster, centr, size, c
def method(k, clusters, centr):
    plt.figure(figsize = (10, 10))
    for i in range(k):
        for j in range(len(clusters[i])):
            plt.scatter(clusters[i][j][0], clusters[i][j][1], color= COLORS[i], s=10)
        plt.scatter(round(centr[i][0], 3), round(centr[i][1], 3), color= COLORS[i],  s=50)
    plt.grid(True)   
    plt.show()   
    
for i in range(1, 8):
    cluster, centr, size_means, c_means = k_means(n,i)
    print(method(i, cluster, centr))
```

### –ó–∞–¥–∞–Ω–∏–µ 4
–†–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º—ã k-medians. –û—Ç–æ–±—Ä–∞–∑–∏—Ç—å –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ –∫–ª–∞—Å—Ç–µ—Ä—ã, –∫–∞–∂–¥—ã–π –∫–ª–∞—Å—Ç–µ—Ä –≤—ã–¥–µ–ª–∏—Ç—å —Ä–∞–∑–Ω—ã–º —Ü–≤–µ—Ç–æ–º –∏–ª–∏ –º–∞—Ä–∫–µ—Ä–æ–º,
–æ—Ç–º–µ—Ç–∏—Ç—å —Ü–µ–Ω—Ç—Ä–æ–∏–¥—ã.

```python
def manhattan (x, y):
    h2 = abs(x[0] - y[0]) + abs(x[1] - y[1])
    return h2
def kmedian (n, k):
    cluster = {}
    cluster = [[] for i in range (k)]
    centr = {}
    size, c = [], []
    index_centr = [10, 25, 45, 50, 75, 90, 100]
    for i in range (k):
        centr[i] = [XY[index_centr[i]][0], XY[index_centr[i]][1]] 
    pred_centr = copy.deepcopy(centr)
    change = True
    while (change):
        change = False
        for i in range(101):
            distance = []
            for j in range (k):
                distance.append(manhattan(XY[i], centr[j]))
            cluster[np.argmin(distance)].append(XY[i])
        for i in range(k):
            if (len(cluster[i]) != 0): centr[i]=[np.median(np.array(cluster[i])[:,0]),np.median(np.array(cluster[i])[:,1])]
            if (pred_centr == centr): change = True
        pred_centr = copy.deepcopy(centr)
    for i in range (k):
        size.append(len(cluster[i]))
    for i in range (k):
        c.append([np.round(centr[i][0], 3), np.round(centr[i][1], 3)])
    return cluster, centr, size, c   
for i in range(1, 8):
    cluster, centr, size_m, c_m = kmedian(n,i)
    print(method(i, cluster, centr))
```

### –ó–∞–¥–∞–Ω–∏–µ 6
–° –ø–æ–º–æ—â—å—é –º–µ—Ç–æ–¥–∞ —Å–∏–ª—É—ç—Ç–æ–≤ (–≤–∞—Ä–∏–∞–Ω—Ç 7) –≤—ã—è–≤–∏—Ç—å –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –º–µ—Ç–æ–¥–∞
–æ–ø—Ç–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–ª–∞—Å—Ç–µ—Ä–æ–≤.

```python
def siluet (m,s,k):
    plt.figure(figsize = (10, 10))
    for i in range(2, 8):
        A, B = [], []
        S = 0
        cluster = m(n, i)[0]
        for j in range(i):
            for k in range(len(cluster[j])):
                b=[]
                a=0
                for l in range(len(cluster[j])):
                    if k != l:
                        a = a + s(cluster[j][k], cluster[j][l])
                A.append(a/(len(cluster[j])-1))
                for z in range(i):
                    if j!=z:
                        e=0
                        for p in range(len(cluster[z])):
                            e = e + s(cluster[j][k], cluster[z][p])
                        b.append(e/len(cluster[z]))
                B.append(min(b))
        for q in range(n):
            S = S + (((B[q] - A[q]) / max(B[q],A[q]))/n)
        plt.scatter(1, 0)
        plt.scatter(i, S)
        plt.grid(True)
        plt.show
print("1 –≥—Ä–∞—Ñ–∏–∫ –¥–ª—è kmeans, 2 –¥–ª—è kmedian")
print(siluet(k_means, evclid, k))
print(siluet(kmedian, manhattan, k))    
```

### –ó–∞–¥–∞–Ω–∏–µ 7
 –†–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º –ø–æ–∏—Å–∫–∞ —Å–≥—É—â–µ–Ω–∏–π –ø—Ä–∏ —Ä–∞–∑–ª–∏—á–Ω—ã—Ö —Ä–∞–¥–∏—É—Å–∞—Ö
ùëÖ, –æ—Ç–æ–±—Ä–∞–∑–∏—Ç—å –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ –∫–ª–∞—Å—Ç–µ—Ä—ã, –≤—ã–¥–µ–ª–∏—Ç—å –∫–∞–∂–¥—ã–π –∫–ª–∞—Å—Ç–µ—Ä
—Ä–∞–∑–Ω—ã–º —Ü–≤–µ—Ç–æ–º –∏–ª–∏ –º–∞—Ä–∫–µ—Ä–æ–º, –æ—Ç–º–µ—Ç–∏—Ç—å —Ü–µ–Ω—Ç—Ä–æ–∏–¥—ã. –†–µ–∑—É–ª—å—Ç–∞—Ç—ã
–ø—Ä–µ–¥—Å—Ç–∞–≤–∏—Ç—å –≤ –≤–∏–¥–µ —Ç–∞–±–ª–∏—Ü—ã:

```python
R_max = -10000
R_min = 10000
def FOREL(R):
    u = [[XY[i][0], XY[i][1]] for i in range (n)]
    cluster, centr = {}, {}
    k = 0
    while(len(u) > 0):
        change = True
        xc = u[0]
        while change:
            u0, u1 = 0, 0
            cluster[k]=[]
            for i in range(len(u)):
                if evclid(xc, u[i]) <= R:
                    u0 += u[i][0]
                    u1 += u[i][1]
                    cluster[k].append(u[i])
            new_xc = xc
            xc = [1/(len(cluster[k])) * u0, 1 / len(cluster[k]) * u1]
    
            if xc == new_xc:
                centr[k] = xc
                change = False             
        for i in cluster[k]:
            if i in u:
                u.remove(i)
        k = k + 1
    size = []
    c = []
    for i in range (k):
        size.append(len(cluster[i]))
    for i in range (k):
        c.append([round(centr[i][0], 4), round(centr[i][1], 4)])
    return k, cluster, centr, size, c
for i in range (n-1):
    for j in range (i+1, n):
            d = evclid(XY[i],XY[j])
            if d > 0:
                if R_max < d:
                    R_max = d
                if R_min > d:
                    R_min = d
print('R_max=', R_max)
print('R_min=', R_min)
def forel(k, clusters, centr):
    plt.figure(figsize = (10,10))
    for i in range(k):
        for j in range(len(clusters[i])):
            plt.scatter(clusters[i][j][0], clusters[i][j][1], color= COLORS[i], s=10)
        plt.scatter(round(centr[i][0], 4), round(centr[i][1], 4), s=50, color= COLORS[i])
    plt.grid(True)
    plt.show()
R = [1.5, 2, 2.5, 3, 5.5]
for i in range(len(R)):
    k, cluster, centr, size, c = FOREL(R[i])
    forel(k, cluster, centr) 
size, c = [], []
for i in range(len(R)):
    size.append(FOREL(R[i])[3])
for i in range(len(R)):
    c.append(FOREL(R[i])[4])    
count3 = pd.DataFrame({'R': R, "–¶–µ–Ω—Ç—Ä—ã": c, '—Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å': size})
print(count3)
```

## –í—ã–≤–æ–¥—ã

–≤ —á–∞—Å—Ç–Ω–æ—Å—Ç–∏, –º–µ—Ç–æ–¥–∞ k-means –∏ –º–µ—Ç–æ–¥–∞ –ø–æ–∏—Å–∫–∞ —Å–≥—É—â–µ–Ω–∏–π.
