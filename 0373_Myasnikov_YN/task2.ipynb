## Основные теоретические положения

Выборка - совокупность n объектов из генеральной совокупности N. 

Интервальный вариационный ряд - промежуток изменения значений выборки $(x_{min}; x_{max})$, разбитый на частичные интервалы.

Для расчета количества интервалов применяется формула Стерджа: $k=1+\log_2 n$.

Размахом выборки называется величина $x_{max}-x_{min}$.

Длина интервала считается по формуле: $h=\frac{R}{k}$.

Выборочное среднее - среднее арифметическое выборочной совокупности $\bar x_в = \frac{1}{n}\sum{x_i \cdot n_i}$

Выборочная дисперсия - мера разброса значений случайной величины относительно её математического ожидания 
$\sigma^2 = \frac{1}{n}\sum{n_i \cdot (x_i-\bar x_в)^2}$.

Мода интервального ряда - число в интервале с наибольшей частотой.
$$M_o = x_{M_o}+h_{M_o} \cdot \frac{n_{M_o}-n_{M_o-1}}{(n_{M_o}-n_{M_o-1})+(n_{M_o}-n_{M_o+1})},$$
где $x_{M_o}$ - левый край интервала с наибольшей частотой, $n_{M_o}$ - частота модального интервала, $n_{M_o+1}$  и $n_{M_o-1}$ - частоты двух соседних интервалов, последующего и предыдущего.

Медиана интервального ряда - число в интервале, для которого накопленная частота составляет половину или больше от всей суммы частот, а предыдущая накопленная частота меньше половины от всей суммы частот. 

$$M_e = x_{M_e}+\frac{h_{M_e}}{p_{M_e}^{*}}\cdot(0.5-p_{M_e-1}^{\sum}),$$
где $x_{M_е}$ - левый край интервала с наибольшей частотой, $p_{M_e-1}^{\sum}$ - накопленная относительная частота медианного интервала

Коэффициент вариации(относительное отклонение) - мера колеблемости изучаемого признака относительно выборочной средней. $V^* = \frac{\sigma_в}{|\bar x_в|}\cdot 100\%$

Условные варианты - варианты, определяемые по неравенству:
$$u_i=\frac{x_i-C}{h},$$ где С - ложный нуль(варианта, расположенная в середине вариационного ряда). 

Выборочный коэффициент ассиметрии: $a_s^* = \frac{\mu_3^*}{\sigma_в^3}$

Выборочный коэффициент эксцесса: $\varepsilon_k^* = \frac{\mu_4^*}{\sigma_в^4} - 3$

Статистическая гипотеза - любое предположение о генеральной совокупности, проверяемое по выборке.

Статистический критерий проверки основной гипотезы - правило, по которому принимается решение, принять гипотезу или нет.

Критическая область - совокупность значений критерия, при которых основную гипотезу отвергают.

Функциональная зависимость - каждому значению одной переменной соответствует единственное значение другой пременной.

Статистическая зависимость - изменение одной из величин влечет изменение распределения другой.

Корреляционная зависимость - изменение одной величины приводит к изменению среднего значения другой.

Условное среднее - среднее арифметическое тех значений случайных величин $X(Y)$, которые соотвестсвуют значению случайной величины $Y = y(X = x)$:
$$\bar x_{y} = \frac{\sum{x_i \cdot n_{x_i y_i}}}{\sum{n_{x_i y_i}}}$$

Межгрупповое среднее: $\overline{xy} = \frac{1}{n} \cdot \sum{\sum{n_{x_i y_j} \cdot x_i \cdot y_j}}$

Уравнениz регрессии: $E(Y|x) = f(x)$  $E(X|y) = f(y)$

Выборочные уравнения регрессии: 
$$\bar{y_x} = f^*(x)$$,  
$$\bar{x_y} = f^*(y)$$

Метод наименьших квадратов:
Задача - свести следующую функцию к минимуму 
$$F(a_1 \dots a_m) = \sum{(y_i - f(x_i; a_1 \dots a_m))^2} -> min$$

Следует приравнять частные производные к нулю:

$$\begin{equation*}
 \begin{cases}
   \frac{dF}{d a_1} = 0
   \\
   \dots
   \\
   \frac{dF}{d a_m} = 0
 \end{cases}
\end{equation*}$$

Решив вышеизложенную систему, получим искомые коэффициенты

Уравнения средней квадратичной регрессии:
$$\bar{y_x} = r_{yx} \cdot \frac{\sigma_y}{\sigma_x}(x - \bar{x}) + \bar{y}$$
$$\bar{x_y} = r_{xy} \cdot \frac{\sigma_x}{\sigma_y}(y - \bar{y}) + \bar{x},$$ 
где $r_{yx} = r_{xy}$ - статистическая оценки коэффициента корреляции

Статистическая оценки коэффициента корреляции через условные варианты: $$r_в = \frac{\frac{1}{n} \cdot \sum{\sum{n_{uv} \cdot u_i \cdot v_j - \bar{u} \cdot \bar{v}}}}{\sigma_u  \cdot \sigma_v}$$

$U_j = \sum{n_{uv} \cdot u_i}$
$V_i = \sum{n_{uv} \cdot v_j}$

Корреляционное отношение:
$\eta_{yx} = \frac{\sigma_{\bar{y} x}}{\sigma_y},$

где $\sigma_{\bar{y} x} = \sqrt{\frac{\sum{n_x \cdot (\bar{y_x} - \bar{y})^2}}{n}}$ , а $\sigma_y = \sqrt{\frac{\sum{n_y \cdot (y - \bar{y})^2}}{n}}$

Доверительный интервал для коэффициента корреляции:
$z = arth(r_в)$, 

тогда искомый доверительный интервал для $z$ равен
$$(z - \frac{t}{\sqrt{n-3}}; z + \frac{t}{\sqrt{n-3}}),$$
где t определяется как $\Phi(t) = \frac{\gamma}{2}$

Доверительный интервал для коэффициента корреляции:
$$(th z_l; th z_r)$$

Проверка гипотезы:
$H_0: r_в = 0$
$H_0: r_в \neq 0$
Наблюдаемое значение критерия считается следующим образом:
$$T_{набл} = r_в \cdot \frac{\sqrt{n-2}}{\sqrt{1-r^2_в}}$$

Критическое значение определяется следующим образом:
$T_{крит} = t(\alpha, df)$ при $df = n-2$

Область принятия гипотезы: $|T_{набл} < T_{крит}|$


Показатели качества(адекватности) модели регрессии:

Теоретический коэффициент детерминации представляет собой отношение объясненной дисперсии признака-результата к общей дисерсии результативного признака
$$R^2_y = \frac{\frac{1}{n} \cdot \sum{(y_{x_i} - \overline{y})^2}}{\sigma^2_y}$$

Средняя квадратическая ошибка - среднее квадратичное отклонение наблюдаемых значений признака от теоретических значений Использование модели регрессии является целесообразным при средней квадратической ошибке меньше СКО
$$S_{\varepsilon y} = \sqrt{\frac{1}{n - l} \cdot \sum{\varepsilon^2_{y_i}}},$$

где $l$ - число параметров в модели регрессии, а $\varepsilon_{y_i} = y_i - y_{x_i}$

Средняя ошибка аппроксимации (приближения):
$$A_y = \frac{1}{n} \cdot \sum{|\frac{\varepsilon_y}{y_i}|} \cdot 100% $$
Из значения меньше 7% следует хорошее качество модели
