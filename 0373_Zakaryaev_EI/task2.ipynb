###–ó–∞–¥–∞–Ω–∏–µ 1
–ü—Ä–æ–≤–µ—Å—Ç–∏ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫—É—é –æ–±—Ä–∞–±–æ—Ç–∫—É –≤—Ç–æ—Ä–æ–π –≤—ã–±–æ—Ä–∫–∏ –≤ –æ–±—ä–µ–º–µ –ø—Ä–∞–∫—Ç–∏—á–µ—Å–∫–æ–π —Ä–∞–±–æ—Ç—ã ‚Ññ1 —Å —Ü–µ–ª—å—é –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Ç–æ—á–µ—á–Ω—ã—Ö —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏—Ö
–æ—Ü–µ–Ω–æ–∫ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∏—Å—Å–ª–µ–¥—É–µ–º–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞ (–º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –æ–∂–∏–¥–∞–Ω–∏—è, –¥–∏—Å–ø–µ—Ä—Å–∏–∏, —Å—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–≥–æ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è,
–∞—Å–∏–º–º–µ—Ç—Ä–∏–∏, —ç–∫—Å—Ü–µ—Å—Å–∞, –º–æ–¥—ã, –º–µ–¥–∏–∞–Ω—ã –∏ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞ –≤–∞—Ä–∏–∞—Ü–∏–∏)

```python
def parseToFloat(string):
  return float(string.replace("$", "").replace("M", ""))

sample = pd.read_csv("/content/moviesSample.csv")
X = 'rating'
Y = 'gross_earn'
N = len(sample)

sample
```

```python
for i, row in sample.iterrows():
  sample.loc[i, Y] = parseToFloat(row[Y])

sample
```

```python
def rangedSample(priznak):
  res = sample.sort_values(priznak)
  res = res.reset_index()
  del(res['index'])
  return res

def getRelFreq(absFreq):
  return absFreq / N

def addToVarSeries(var, varSeries):
  varSeries.loc[len(varSeries)] = {'var': var, 'absFreq': 1, 'relFreq': getRelFreq(1)}

def increaseVarSeriesAbsFreq(var, varSeries):
  varSeries.loc[varSeries['var'] == var, 'absFreq'] += 1
  updateVarSeriesRelFreq(var, varSeries)

def updateVarSeriesRelFreq(var, varSeries):
  varSeries.loc[varSeries['var'] == var, 'relFreq'] = getRelFreq(varSeries.loc[varSeries['var'] == var, 'absFreq'])

def varExists(var, series):
  return (series['var'] == var).any()

def varSeries(priznak):
  varSeries = pd.DataFrame(columns=['var', 'absFreq', 'relFreq'])
  for index, row in rangedSample(priznak).iterrows():
    var = row[priznak]
    if varExists(var, varSeries):
      increaseVarSeriesAbsFreq(var, varSeries)
    else:
      addToVarSeries(var, varSeries)
  return varSeries

def addToIntSeries(var, absFreq, relFreq, intSeries):
  mask = (intSeries['start'] < var) & (var <= intSeries['end'])
  intSeries.loc[mask, 'absFreq'] += absFreq
  intSeries.loc[mask, 'relFreq'] += relFreq

def addIntervalToIntSeries(start, end, intSeries):
  intSeries.loc[len(intSeries)] = {'start': start, 'end': end, 'mid': (start + end) / 2, 'absFreq': 0, 'relFreq': 0}

def splitToIntervals(priznak):
  k = round(1 + np.log2(N))
  max = rangedSample(priznak).loc[len(rangedSample(priznak)) - 1, priznak]
  min = rangedSample(priznak).loc[0, priznak]
  R = max - min
  h = R/k
  res = []
  for i in range(0, k):
    start = min + h * i
    end = min + h * (i + 1)
    if i == 0:
      start -= 0.01
    res += [(start, end)]
  return res

def intSeries(priznak):
  intSeries = pd.DataFrame(columns=['start','end','mid','absFreq','relFreq'])
  
  for i in splitToIntervals(priznak):
    addIntervalToIntSeries(i[0], i[1], intSeries)
  
  for index, row in varSeries(priznak).iterrows():
    addToIntSeries(float(row['var']), float(row['absFreq']), float(row['relFreq']), intSeries)
  
  intSeries.loc[:,'accumRelFreq'] = 0
  intSeries.loc[:,'accumAbsFreq'] = 0
  accumRelFreq = 0
  accumAbsFreq = 0

  for index, row in intSeries.iterrows():
    accumRelFreq += row['relFreq']
    intSeries.loc[index,'accumRelFreq'] = accumRelFreq
    accumAbsFreq += row['absFreq']
    intSeries.loc[index,'accumAbsFreq'] = accumAbsFreq

  return intSeries

intSeries(Y)
```

```python
def mean(priznak):
  xv = 0
  for index, row in varSeries(priznak).iterrows():
    xv += row['var'] * row['relFreq']
  return xv

def centerMoment(i, priznak):
  moment = 0
  avg = mean(priznak)
  for index, row in varSeries(priznak).iterrows():
    moment += (row['var'] - avg)**i * row['relFreq']
  return moment

def disp(priznak):
  return centerMoment(2, priznak)

def sko(priznak):
  return disp(priznak)**0.5

def ispDisp(priznak):
  return disp(priznak) * N / (N-1)

def ispSko(priznak):
  return ispDisp(priznak)**0.5

def assym(priznak):
  return centerMoment(3, priznak) / sko(priznak)**3

def exc(priznak):
  return centerMoment(4, priznak) / disp(priznak)**2 - 3

def moda(priznak):
  moda = 0
  freq = -1
  for index, row in varSeries(priznak).iterrows():
    if(row['absFreq'] > freq):
      freq = row['absFreq']
      moda = row['var']
  return moda

def median(priznak):
  mediana = rangedSample(priznak).loc[len(varSeries(priznak)) // 2, priznak]
  return mediana
    
def koefVar(priznak):
  return sko(priznak) / abs(mean(priznak))

print("–°—Ä–µ–¥–Ω–µ–µ –≤—ã–±–æ—Ä–æ—á–Ω–æ–µ Y: ", mean(Y))
print("–í—ã–±–æ—Ä–æ—á–Ω–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è Y: ", disp(Y))
print("–í—ã–±–æ—Ä–æ—á–Ω–æ–µ –°–ö–û Y: ", sko(Y))
print("–í—ã–±–æ—Ä–æ—á–Ω–∞—è –∏—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è Y: ", ispDisp(Y))
print("–í—ã–±–æ—Ä–æ—á–Ω–æ–µ –∏—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω–æ–µ –°–ö–û Y: ", ispSko(Y))
print("–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∞—Å—Å–∏–º–µ—Ç—Ä–∏–∏ Y:", assym(Y))
print("–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç —ç–∫—Å—Ü–µ—Å—Å–∞ Y:", exc(Y))
print("–ú–æ–¥–∞ Y:", moda(X))
print("–ú–µ–¥–∏–∞–Ω–∞ Y:", median(Y))
print("–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –≤–∞—Ä–∏–∞—Ü–∏–∏ Y:", koefVar(Y))
```

```python
def countVariantsInInterval(startX, endX, startY, endY):
  count = 0
  for i, row in sample.iterrows():
    if(startX < row[X] and row[X] <= endX):
      if(startY < row[Y] and row[Y] <= endY):
        count += 1
  return count

def format(_float):
  return round(_float, 2)

xIntervals = splitToIntervals(X)
yIntervals = splitToIntervals(Y)

intSeries = pd.DataFrame(columns=["Y\X"] + [f"({format(i[0])};{format(i[1])}]" for i in xIntervals] + ["n_y"])
for y in yIntervals:
  counts = [countVariantsInInterval(x[0], x[1], y[0], y[1]) for x in xIntervals]
  intSeries.loc[len(intSeries)] = [f"({format(y[0])};{format(y[1])}]"] + counts + [sum(counts)]

counts = []
for x in xIntervals:
  counts = counts + [sum([countVariantsInInterval(x[0], x[1], y[0], y[1]) for y in yIntervals])]

intSeries.loc[len(intSeries)] = ["n_x"] + counts + [sum(counts)]

intSeries
```

###–ó–∞–¥–∞–Ω–∏–µ 3

Y: –ü–µ—Ä–≤—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª $(0.01;117.1]$ —è–≤–ª—è–µ—Ç—Å—è –º–æ–¥–∞–ª—å–Ω—ã–º, —Ç–æ–≥–¥–∞ –æ–Ω –≤—ã—Å—Ç—É–ø–∞–µ—Ç –≤ –∫–∞—á–µ—Å—Ç–≤–µ $C_y$ –ø—Ä–∏ –ø–æ–¥—Å—á–µ—Ç–µ —É—Å–ª–æ–≤–Ω—ã—Ö –≤–∞—Ä–∏–∞–Ω—Ç. –î–ª–∏–Ω—ã –∏–Ω—Ç–µ—Ä–≤–∞–ª–æ–≤ –æ–¥–∏–Ω–∞–∫–æ–≤—ã, —Ç–æ–≥–¥–∞ –∏–º–µ–µ–º —É—Å–ª–æ–≤–Ω—ã–µ –≤–∞—Ä–∏–∞–Ω—Ç—ã $v=\{0, 1, 2, 3, 4, 5, 6, 7\}$

X: –ú–æ–¥–∞–ª—å–Ω—ã–º —è–≤–ª—è–µ—Ç—Å—è –∏–Ω—Ç–µ—Ä–≤–∞–ª $(6.9;7.6]$. –î–ª–∏–Ω—ã –∏–Ω—Ç–µ—Ä–≤–∞–ª–æ–≤ –æ–¥–∏–Ω–∞–∫–æ–≤—ã. –¢–æ–≥–¥–∞ –∏–º–µ–µ–º —É—Å–ª–æ–≤–Ω—ã–µ –≤–∞—Ä–∏–∞–Ω—Ç—ã $u=\{-5, -4 , -3, -2, -1, 0, 1, 2\}$

```python
uslSeries = pd.DataFrame(
    index=["0", "1", "2", "3", "4", "5", "6", "7", "n_u", "V_j", "u_jV_j"],
    columns=["-5", "-4", "-3", "-2", "-1", "0", "1", "2", "n_v", "U_i", "v_iU_i"])

uslSeries
```

–°–Ω–∞—á–∞–ª–∞ –Ω—É–∂–Ω–æ —Å–∫–æ–ø–∏—Ä–æ–≤–∞—Ç—å –∑–Ω–∞—á–µ–Ω–∏—è –≤ –Ω–æ–≤—É—é —Ç–∞–±–ª–∏—Ü—É

```python
for i in range(intSeries.shape[0]):
  copied = intSeries.iloc[i].values.tolist() + [np.NaN, np.NaN]
  copied.pop(0)
  uslSeries.iloc[i] = copied

uslSeries
```

–ü–æ–¥—Å—á–µ—Ç –∑–Ω–∞—á–µ–Ω–∏–π $V_j$ –∏ $u_jV_j$

```python
def amountOfUslVariants():
  return len(uslSeries) - 3

vFirst = 0

for v in uslSeries:
  if v == "n_v":
    break
  col = uslSeries[v].values.tolist()
  del col[-1]
  del col[-1]
  del col[-1]
  for i in range(len(col)):
    col[i] = col[i] * (vFirst + i)
  v_j = sum(col)
  u_jv_j = int(v) * v_j
  uslSeries[v]["V_j"] = v_j
  uslSeries[v]["u_jV_j"] = u_jv_j

uslSeries
```

–ü–æ–¥—Å—á–µ—Ç –∑–Ω–∞—á–µ–Ω–∏–π $U_i$ –∏ $v_iU_i$

```python
uFirst = -5

for i in range(amountOfUslVariants()):
  row = uslSeries.iloc[i].values.tolist()
  del row[-1]
  del row[-1]
  del row[-1]
  for j in range(len(row)):
    row[j] = row[j] * (uFirst + j)
  u_i = sum(row)
  v_iu_i = i * u_i
  uslSeries.iloc[i]["U_i"] = u_i
  uslSeries.iloc[i]["v_iU_i"] = v_iu_i

uslSeries
```

–ö–æ–Ω—Ç—Ä–æ–ª—å $\sum{u_jV_j}=-3-3+7+6=\sum{v_iU_i}=-6+6+7=7$

```python
ujVjs = uslSeries.loc["u_jV_j"].values.tolist()
del ujVjs[-1]
del ujVjs[-1]
del ujVjs[-1]
uslSeries["v_iU_i"]["u_jV_j"] = sum(ujVjs)
uslSeries
```

###–ó–∞–¥–∞–Ω–∏–µ 4

–í—ã—á–∏—Å–ª–µ–Ω–∏–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏

```python
avgU = 0
for i in range(amountOfUslVariants()):
  avgU += uslSeries[str(uFirst + i)]["n_u"] * (uFirst + i)
avgU /= N

print("u —Å—Ä–µ–¥–Ω–µ–µ:", avgU)

avgV = 0
for i in range(amountOfUslVariants()):
  avgV += uslSeries["n_v"][str(vFirst + i)] * (vFirst + i)
avgV /= N

print("v —Å—Ä–µ–¥–Ω–µ–µ:", avgV)

skoU = 0
for i in range(amountOfUslVariants()):
  skoU += uslSeries[str(uFirst + i)]["n_u"]*(uFirst + i - avgU)**2
skoU = np.sqrt(skoU/amountOfUslVariants())

print("–°–ö–û u:",skoU)

skoV = 0
for i in range(amountOfUslVariants()):
  skoV += uslSeries["n_v"][str(vFirst + i)]*(vFirst + i - avgU)**2
skoV = np.sqrt(skoV/amountOfUslVariants())

print("–°–ö–û v:",skoV)

r = ((1/N)*uslSeries["v_iU_i"]["u_jV_j"] - avgU * avgV)/(skoU * skoV)

print("\n–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏:", r)
```

–ü–æ–ª—É—á–µ–Ω –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ $r_–≤$ –±–ª–∏–∑–∫–∏–π –∫ –Ω—É–ª—é. –≠—Ç–æ –æ–∑–Ω–∞—á–∞–µ—Ç, —á—Ç–æ –Ω–∞ –ø–æ–ª—É—á–µ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ –ª–∏–Ω–µ–π–Ω–æ–π –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –Ω–µ –Ω–∞–±–ª—é–¥–∞–µ—Ç—Å—è.

###–ó–∞–¥–∞–Ω–∏–µ 5

–ù–∞—Ö–æ–∂–¥–µ–Ω–∏–µ –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã—Ö –∏–Ω—Ç–µ—Ä–≤–∞–ª–æ–≤ –æ—Ü–µ–Ω–∫–∏ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ 

```python
t95 = 1.9821734832574511 # –ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –°—Ç—å—é–¥–µ–Ω—Ç–∞ t(0.95, 109)
t99 = 2.622120060014357 # –ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –°—Ç—å—é–¥–µ–Ω—Ç–∞ t(0.99, 109)

z = math.atanh(r)
int95 = (math.tanh(z - ( t95/np.sqrt(N - 3))), math.tanh(z + ( t95/np.sqrt(N - 3))))
int99 = (math.tanh(z - ( t99/np.sqrt(N - 3))), math.tanh(z + ( t99/np.sqrt(N - 3))))

print("–î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª –∫–æ—ç—Ñ—Ñ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –¥–ª—è 95%:", int95)
print("–î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª –∫–æ—ç—Ñ—Ñ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –¥–ª—è 99%:", int99)
```

###–ó–∞–¥–∞–Ω–∏–µ 6

–ü—Ä–æ–≤–µ—Ä–∫–∞ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –≥–∏–ø–æ—Ç–µ–∑—ã –æ —Ä–∞–≤–µ–Ω—Å—Ç–≤–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ $r$ –Ω—É–ª—é –ø—Ä–∏ –∑–∞–¥–∞–Ω–Ω–æ–º —É—Ä–æ–≤–Ω–µ –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ $\alpha = 0.05$

```python
tNabl = r * np.sqrt(N - 2) / np.sqrt(1 - r**2)
print("T_–Ω–∞–±–ª =", tNabl)
tCrit = 1.659219 # t(0.05, 109-2=107)
print("T_–∫—Ä–∏—Ç =", tCrit)
```

–û–±–ª–∞—Å—Ç—å –ø—Ä–∏–Ω—è—Ç–∏—è –≥–∏–ø–æ—Ç–µ–∑—ã: $|T_{–Ω–∞–±–ª}|<T_{–∫—Ä–∏—Ç}$

$0.085=|T_{–Ω–∞–±–ª}|<T_{–∫—Ä–∏—Ç}=1.66$

–°–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ, –≥–∏–ø–æ—Ç–µ–∑–∞ –æ —Ä–∞–≤–µ–Ω—Å—Ç–≤–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ $r$ –Ω—É–ª—é –ø—Ä–∏–Ω—è—Ç–∞.


###–ó–∞–¥–∞–Ω–∏–µ 7

```python
skoX = sko(X)
skoY = sko(Y)
avgX = mean(X)
avgY = mean(Y)

a1 = r*(skoY/skoX)
a2 = r*(skoX/skoY)

b1 = -a1 * avgX + avgY
b2 = -a2 * avgY + avgX

print(a1, b1)
print(a2, b2)
```
